# план-факт анализ
В первую очередь нужно сфокусироваться на отклонениях от плана, то есть определить, на сколько фактические данные отличаются от плановых значений. 

Для этого нужно **вычесть из фактических показателей запланированные**. Так вы рассчитаете абсолютные отклонения от плана. 

**Абсолютное отклонение в финансах** — это разница между запланированными и фактическими данными, которые отражают один и тот же финансовый показатель. Выражается эта разница в денежных единицах.

<img src='images/00026.png' width=800><br> 

**Относительные отклонения в финансах** — это соотношение одного абсолютного показателя к другому. Чаще всего выражается в процентах. 

Если финансовый аналитик хочет выяснить, выполнен ли план, — нужно **фактическое значение разделить на плановое и умножить на 100**. Если результат получается больше 100% — значит, план перевыполнен; меньше — недовыполнен. 
В случае сети кофеен «Элементарно» план по продаже «кофе с собой» выполнен всего на 86%: `1 782 870 / 2065252*100 = 86%`

Чтобы посчитать, на сколько в процентах изменилась выручка по продаже готового кофе, нужно взять данные абсолютного отклонения и разделить на плановое значение выручки, а затем умножить на 100:
`282382/2065252*100=13.6%`

Как проверить, повлияла ли сезонность на отклонение по прибыли в июне?

 - Посмотреть план выручки по месяцам
Если в бизнесе есть сезонность, необходимо проверить, заложили ли её в план. В случае с кофейнями, если сезонность не учли, очевидно, что отклонения от плана в низкий сезон будут.

 - Посмотреть динамику продаж в прошлом году
Да, с этого и нужно начать исследование сезонности бизнеса.

 - Посмотреть на изменение расходов, которые относятся к себестоимости
Себестоимость, так же как и выручка, зависит от сезонности. Представьте, что кофейня планировала продать 100 порций кофе на вынос, для которых понадобится 100 стаканчиков. Если в итоге продали только 70 порций, то и стаканчиков потребовалось 70. То же самое относится к количеству зерна и молока.


Может сложиться ситуация, что прибыль отклонилась в том числе и из-за изменения цен у поставщиков, а не только из-за сезонности. Для этого нужно провести план-факт анализ себестоимости.

Нужно посмотреть, изменяется ли себестоимость в те же периоды и примерно в том же размере, что и выручка. Но перед тем как смотреть отклонения плана и факта, нужно объединить в одну группу расходы, которые относятся к готовому кофе, — это кофе, молоко и расходные материалы. Какие показатели оказали негативное влияние на прибыль?

 - ВыручкаОплата труда
 - Оплата труда тоже негативно сказалась на прибыли. Чтобы понять, насколько негативно, нужно провести факторный анализ.
 - Аренда, коммунальные платежиЭти расходы негативно повлияли на прибыль. Дальше рассмотрим с вами, почему.
 - Рекламные расходыУвеличение рекламных расходов, безусловно, сказалось на прибыли. Факторный анализ позволит понять, главная ли эта причина отклонения прибыли.

Из план-факт анализа может показаться, что арендная плата и оплата труда положительно отразились на прибыли, так как фактические данные меньше запланированных. Однако это не совсем так. Чтобы понять почему, требуется провести факторный анализ — другой вид финансового анализа.3
