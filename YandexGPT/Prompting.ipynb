# Уточнение контекста и использование примеров
четыре ключевые техники для улучшения промптов: 
 - предоставление контекста, 
 - использование примеров, 
 - введение ограничений и 
 - выбор между открытыми и закрытыми вопросами. 

Эти приёмы помогают вам создавать чёткие, точные запросы и получать ответы, которые соответствуют вашим ожиданиям.
Помните, что успех в промпт-инжиниринге заключается в деталях. Чем лучше вы уточняете задачу, тем больше шансов, что модель даст вам полезный и качественный ответ.
## Provide Context (Предоставление контекста)

Контекст — это информация, которая помогает модели понять, в какой ситуации или для какой цели вы задаёте вопрос. Чем больше деталей вы предоставите, тем более полезный и целевой ответ получите.

Почему контекст важен?
Модель не обладает "знаниями о мире" в момент общения. Она обрабатывает ваш запрос, основываясь только на предоставленной информации. Без контекста модель может неправильно интерпретировать ваш запрос.

Как добавить контекст?

`Укажите, кто вы или кто является целевой аудиторией.`

Пример:
"Представь, что ты специалист по маркетингу. Расскажи, как продвигать бренд через социальные сети."

`Объясните цель вашего запроса.`

Пример:
"Мне нужно описание продукта для интернет-магазина. Продукт: умные часы. Описание должно быть ориентировано на молодёжную аудиторию."

`Добавьте дополнительные детали.`

Пример:
"Напиши текст приглашения на конференцию по теме искусственного интеллекта. Целевая аудитория — разработчики. Тон формальный."

Пример использования контекста:

Промпт без контекста:
"Объясни, что такое блокчейн."

Ответ будет общим и менее целевым.

Промпт с контекстом:
"Объясни, что такое блокчейн, используя простые термины, понятные старшекласснику, изучающему основы информатики."

Ответ будет адаптирован под аудиторию и цель.
 
## Few-shot Prompting (Примеры)
Few-shot prompting — это техника, которая позволяет улучшить понимание модели, предоставляя ей несколько примеров желаемого результата. Это особенно полезно для сложных или творческих задач.

Почему примеры работают?

Модель учится на предоставленных примерах, подстраивая свои ответы под их формат и стиль. Примеры помогают задать "рамки", в которых модель должна работать.

Как использовать примеры?

Дайте 2-3 примера, демонстрирующих желаемый результат.

Пример 1:

Вопрос: Что такое электричество?

Ответ: Электричество — это поток электронов через проводник.
 
Пример 2:

Вопрос: Что такое гравитация?

Ответ: Гравитация — это сила, которая притягивает объекты друг к другу."

Задайте вопрос, используя схожую структуру.

"Теперь ответь на вопрос: Что такое искусственный интеллект?"
 
## Explicit Constraints (Ограничения)
Ограничения помогают уточнить, каким должен быть ответ. Вы можете задать стиль, длину текста, тон, форматирование и даже целевую аудиторию. Это особенно полезно для специфичных задач, например, написания маркетинговых текстов или инструкций.

Какие ограничения можно задавать?
 - Стиль: дружелюбный, официальный, научный, креативный.

Пример:
"Напиши описание смартфона в дружелюбном тоне для молодёжи."

 - Длина: укажите, сколько слов или предложений должно быть в ответе.

Пример:
"Объясни, как работает искусственный интеллект, в двух предложениях."

 - Тон: формальный, неформальный, профессиональный.

Пример:
"Напиши письмо клиенту в профессиональном тоне, предлагая скидку на покупку."

- Форматирование: списки, таблицы, абзацы.

Пример:
"Напиши список из 5 преимуществ электрических автомобилей."

Пример использования ограничений:
"Напиши статью о влиянии глобального потепления на природу. Текст должен быть длиной до 150 слов, в научно-популярном стиле, без сложных терминов."
 
## Open-Ended vs. Close-Ended Prompts (Открытые и закрытые вопросы)
Выбор между открытыми и закрытыми запросами зависит от того, какой тип ответа вы хотите получить.

### Открытые вопросы
Открытые вопросы предназначены для получения развёрнутого ответа.

Пример:
"Какие способы решения проблемы загрязнения океанов существуют?"
Ответ будет включать несколько вариантов или идей.

### Закрытые вопросы
Закрытые вопросы формулируются так, чтобы получить конкретный, краткий ответ.

Пример:
"Сколько процентов Земли покрыто океанами?"
Ответ будет точным и фактическим.

Как выбрать правильный тип вопроса?
 - Используйте открытые вопросы для генерации идей, анализа или изучения темы.
 - Используйте закрытые вопросы для получения конкретной информации или фактов.
Примеры:
 - Открытый: "Какую пользу приносит регулярное чтение книг?"
 - Ответ: список преимуществ или объяснение.

 - Закрытый: "Сколько часов в день рекомендуется читать для пользы здоровья?"
 - Ответ: точное количество часов.

# Углублённые техники и управление диалогом

углублённые техники, такие как 
 - пошаговые инструкции, 
 - многошаговый диалог и 
 - итеративное уточнение.

## Step-by-Step Instructions (Пошаговые инструкции)
Одной из самых мощных техник является разбиение сложных задач на этапы. Пошаговые инструкции помогают модели отвечать последовательно и структурировано, избегая пропуска важных деталей.

Почему это важно?<br>
Когда задача комплексная, модель может растеряться или ответить слишком поверхностно. Пошаговые инструкции помогают разбить задачу на управляемые части, делая ответ логичным и структурированным.

Как применять пошаговые инструкции?<br>
Чётко укажите, что задача должна быть выполнена по шагам.<br>
Пример:<br>
"Объясни, как создать бюджет для личных финансов, поэтапно. Опиши каждый шаг подробно."

Используйте маркеры (пункты, нумерацию), чтобы явно указать структуру.<br>
Пример:<br>
"Напиши руководство по установке программного обеспечения. Раздели процесс на 5 шагов."<br>

Пример использования:<br>
"Объясни, как приготовить пасту карбонара, пошагово:"<br>
Ответ:<br>
```
Подготовьте ингредиенты: спагетти, бекон, яйца, сыр пармезан, соль и перец.<br>
Отварите спагетти до состояния al dente.
Обжарьте бекон до золотистой корочки.
Смешайте яйца и натёртый пармезан в миске.
Соедините всё вместе, добавив яйца с сыром в горячую пасту. Перемешайте.
```
Таким образом, модель понимает, что от неё требуется последовательный и логичный ответ.
 
 
## Multi-turn Prompting (Многошаговый диалог)
Многошаговый диалог — это техника, при которой вы постепенно уточняете запрос, задавая дополнительные вопросы. Это особенно полезно для сложных задач, где детали уточняются шаг за шагом.

Почему это работает?<br>
Модель не всегда может выдать полный и точный ответ с первого раза. Постепенное уточнение запроса позволяет "направлять" модель и добиваться более точных результатов.

Как строить многошаговый диалог?<br>
Начните с общего запроса, чтобы получить начальную информацию.<br>
Пример:<br>
"Расскажи, что такое искусственный интеллект."<br>

Уточните отдельные аспекты, которые вас интересуют.<br>
Пример:<br>
"Теперь поясни, как нейронные сети используются для распознавания изображений."<br>

Продолжайте задавать дополнительные вопросы, чтобы углубиться в тему.<br>
Пример:<br>
"Можешь объяснить, как обучаются такие сети? И какие данные для этого нужны?"<br>

Пример использования:<br>
Вы хотите узнать о создании мобильного приложения:<br>
"Объясни, как создать мобильное приложение."<br>
"Теперь поясни, какие инструменты используются для разработки под iOS."<br>
"Какие навыки нужны разработчику для этой задачи?"<br>
Такой подход позволяет получить последовательный и полный ответ, охватывающий все интересующие вас аспекты.<br>
 
## Iterative Refinement (Итеративное уточнение)
Итеративное уточнение — это процесс постепенной доработки запроса на основе полученного ответа. Если первый ответ модели недостаточно точен или не соответствует ожиданиям, вы можете уточнить запрос и получить лучший результат.

Почему это важно?<br>
Модель не всегда понимает, что именно вы хотите, с первого раза. Итеративное уточнение позволяет "настроить" её на нужный вам результат.
<br>
Как уточнять запросы?<br>
Укажите, что нужно исправить или улучшить в ответе.<br>
Пример:<br>
"Сделай текст более простым и понятным для подростков."<br>

Добавьте новые детали или ограничения.<br>
Пример:<br>
"Расширь ответ, добавив примеры из реальной жизни."<br>

Спросите, если что-то осталось неясным.<br>
Пример:<br>
"Объясни подробнее, что ты имеешь в виду под этим термином."<br>

Пример итеративного уточнения:<br>
Промпт: "Напиши статью о пользе регулярного сна."<br>
Ответ: "Сон важен для здоровья, так как он восстанавливает силы."<br>
Уточнение: "Расширь ответ и добавь научные факты о том, как сон влияет на мозг."<br>
Ответ: "Сон восстанавливает силы, улучшает память и помогает мозгу обработать информацию. Исследования показывают, что 7-9 часов сна в день снижают риск депрессии."<br>
Итеративный процесс помогает добиваться более полного и качественного ответа.<br>
 
 
## Error-Handling Prompts (Обработка ошибок)
Модель может столкнуться с ситуациями, когда у неё недостаточно данных для точного ответа, или она не понимает ваш запрос. Для таких случаев важно заранее предусмотреть, как она должна действовать.

Почему это важно?<br>
Если модель сталкивается с неопределённостью, она может либо дать общий ответ, либо предложить неподходящий вариант. Указание, как действовать в таких ситуациях, позволяет получить более адекватный результат.<br>

Как задать обработку ошибок?<br>
Укажите, что делать, если модель не знает ответа.<br>
Пример:<br>
"Если у тебя недостаточно данных, предложи несколько возможных вариантов."<br>

Попросите уточнить, если что-то непонятно.<br>
Пример:<br>
"Если мой запрос непонятен, уточни его перед тем, как ответить."<br>

Укажите, что делать с недостаточной информацией.<br>
Пример:<br>
"Если ты не можешь ответить, предложи общие рекомендации или уточни, что именно тебе нужно знать."<br>

Пример обработки ошибок:<br>
Промпт: "Расскажи о малоизвестных фактах о древнем городе."<br>
Ответ: "Мне неизвестны такие факты. Возможно, вы могли бы уточнить, о каком городе идёт речь?"<br>
Уточнение: "О древнем городе Помпеи."<br>
Ответ: "Помпеи известны своими археологическими раскопками. Например, город был погребён под вулканическим пеплом в 79 году н.э."<br>
Обработка ошибок делает общение с моделью более гибким и продуктивным.<br>
```
Пошаговые инструкции помогают структурировать задачи и делать ответы логичными.
Многошаговый диалог позволяет глубже изучить тему, уточняя детали шаг за шагом.
Итеративное уточнение даёт возможность дорабатывать запросы и получать более точные результаты.
Обработка ошибок помогает модели справляться с неопределённостью и работать даже в сложных ситуациях.
Каждая из этих техник делает вас более уверенным пользователем промпт-инжиниринга. Практикуйтесь в их использовании: начните с простых задач, а затем переходите к более сложным, применяя пошаговый подход, уточнения и управление ошибками.
```

# Креативность, сравнения и гипотетические сценарии
мы поговорим о творческом подходе в промпт-инжиниринге, таких как 
 - креативные ограничения, 
 - гипотетические сценарии и 
 - работа с тональностью. 
Это поможет вам ещё больше расширить возможности использования модели!

## Creative Constraints (Творческие ограничения)
Одна из самых интересных возможностей работы с моделью — это создание креативных и необычных ответов с помощью заданий с творческими ограничениями. Эта техника позволяет вам получать нестандартные результаты, задавая модели особые рамки для работы.

Стандартные ответы модели обычно основываются на логике и фактах, что хорошо для аналитических задач. Однако, если вам нужен креативный контент, такие как уникальные идеи, художественные тексты или нестандартные решения, важно задавать необычные рамки.

Как использовать творческие ограничения?<br>
Добавляйте необычные элементы.<br>
Пример: "Напиши рассказ о том, как искусственный интеллект помог коту стать космонавтом."<br>

Просите модель изменить стиль или формат.<br>
Пример: "Объясни теорию относительности в стихотворной форме."<br>

Задавайте необычные сценарии.<br>
Пример: "Опиши утопический мир, где все работы выполняют роботы, а люди занимаются только искусством."<br>

Пример:<br>
Промпт: "Напиши статью о преимуществах работы удалённо в виде юмористического монолога комика."<br>
Ответ:<br>
"Работа из дома — это когда ты можешь совмещать утренний зум-звонок с приготовлением омлета, а твои коллеги понятия не имеют, что на тебе сейчас пижама с динозаврами. Но знаете, в этом есть свои минусы: например, иногда кот становится твоим самым надёжным слушателем на совещании."<br>
Совет: экспериментируйте с рамками! Чем интереснее ограничения, тем более необычный и вдохновляющий будет результат.<br>
 
## Comparisons (Сравнения)
Сравнение — это мощный инструмент анализа. Он помогает понять различия и сходства между концепциями, методами или объектами. Сравнительные запросы делают ответы более структурированными и аналитичными.

Сравнение позволяет глубже понять предмет, выявить преимущества и недостатки, а также найти подходящее решение для конкретной ситуации.

Как правильно формулировать запросы для сравнения?
 - Определите, что нужно сравнить.<br>
Пример: "Сравни преимущества и недостатки солнечной и ветровой энергии."

 - Задайте критерии для сравнения.<br>
Пример: "Сравни Python и Java для начинающих программистов с точки зрения лёгкости изучения, производительности и популярности."

 - Попросите итоговый вывод.<br>
Пример: "Какой вид транспорта лучше для экологии: электромобили или велосипеды? Подведи итог."

Пример:<br>
Промпт: "Сравни преимущества и недостатки работы в офисе и удалённой работы."<br>
Ответ:<br>
Работа в офисе:<br>
Преимущества: лёгкость общения с коллегами, доступ к офисным ресурсам, чёткая рабочая среда.<br>
Недостатки: необходимость тратить время на дорогу, меньше гибкости в расписании.<br>
Удалённая работа:<br>
Преимущества: экономия времени, комфорт домашней обстановки, гибкость в графике.<br>
Недостатки: отсутствие личного общения, сложности с концентрацией, необходимость самому организовывать рабочее место.<br>
Итог: выбор зависит от индивидуальных предпочтений и условий работы.<br>

`Совет: используйте сравнения, чтобы принимать обоснованные решения и получать структурированные ответы.`
 
 
## Hypothetical Scenarios (Гипотетические сценарии)
Гипотетические сценарии — это техника, позволяющая создавать предположительные ситуации для анализа, прогнозирования или вдохновения. Они отлично подходят для изучения возможных последствий тех или иных решений.

Такие запросы помогают оценить, как будут развиваться события в определённых условиях. Это полезно для анализа рисков, планирования или мозгового штурма.

**Как формулировать гипотетические сценарии?**
 - Создайте условие.<br>
Пример: "Если бы у каждого человека было по роботу, как бы это изменило рынок труда?"

 - Сформулируйте вопрос о последствиях.<br>
Пример: "Какие экологические проблемы могли бы возникнуть, если бы весь транспорт стал электрическим?"

 - Добавьте детали для глубины анализа.<br>
Пример: "Представь, что весь мир перешёл на 4-дневную рабочую неделю. Как это повлияет на экономику и благополучие людей?"

Пример:Промпт: "Что произойдёт, если все города станут умными и будут полностью управляться искусственным интеллектом?"<br>
Ответ:"В умных городах транспорт будет оптимизирован, сократятся пробки, а уровень преступности снизится благодаря продвинутым системам видеонаблюдения. Однако могут возникнуть этические вопросы о конфиденциальности данных и зависимости от технологий. Кроме того, потребуется пересмотр законодательства для регулирования таких городов."<br>
`Совет: используйте гипотетические сценарии для стимулирования дискуссий и поиска новых идей.`
 
## Focus on Tone (Фокус на тоне)
Тон ответа имеет большое значение для восприятия. Он может быть 
 - официальным, 
 - дружелюбным, 
 - мотивационным, 
 - академическим *или даже*
 - игривым.

Выбор тона зависит от вашей аудитории и целей.

Неверный тон может снизить эффективность коммуникации. Например, сложный академический стиль не подойдёт для подростков, а слишком простой стиль может не удовлетворить профессионалов.

**Как задавать тон?**
 - Укажите стиль текста.<br>
Пример: "Напиши текст в дружелюбном и мотивирующем тоне."

 - Определите целевую аудиторию.<br>
Пример: "Объясни, как работает искусственный интеллект, для детей младшего школьного возраста."

 - Добавьте дополнительные ограничения, если нужно.<br>
Пример: "Напиши описание книги в официальном тоне для презентации на конференции."

Пример: Промпт: "Напиши мотивационное письмо для студентов, чтобы вдохновить их изучать программирование."<br>
Ответ: "Дорогие студенты, программирование — это не просто навык, это инструмент, который открывает перед вами мир безграничных возможностей. Сегодня вы можете начать путь, который приведёт вас к созданию технологий будущего. Не бойтесь ошибок, ведь каждая из них — это шаг к вашему росту. Удачи вам на этом удивительном пути!"<br>
`Совет: экспериментируйте с тоном, чтобы добиться наилучшего эффекта в зависимости от аудитории.`


# Оптимизация промптов и интеграция в реальную работу

## Instruction-based Prompting (Чёткая инструкция)
Чёткие инструкции — основа успешного взаимодействия с моделью. Неправильно сформулированный запрос может привести к неопределённым или несоответствующим ожиданиям ответам.

Чем понятнее и точнее запрос, тем более релевантным будет результат. Чёткие инструкции экономят время и позволяют получить ответы, которые сразу можно использовать в работе.

**Как писать чёткие инструкции?** 
 - Будьте конкретными.<br>
Вместо: "Расскажи мне что-нибудь о маркетинге,"<br>
напишите: "Объясни 5 ключевых стратегий контент-маркетинга для продвижения товаров в социальных сетях."

 - Избегайте двусмысленности.<br>
Уточните, что именно вы хотите получить.<br>
Пример: "Напиши 3 идеи для продвижения продукта. Каждая идея должна содержать заголовок и описание длиной до 50 слов."

 - Делайте запрос пошаговым.<br>
Пример: "Опиши процесс создания бизнес-плана, выделяя каждый шаг в отдельный пункт с кратким пояснением."

Пример:<br>
Промпт: "Создай список из 5 экологичных способов сократить расход воды в домашних условиях. Каждое предложение должно быть максимум 20 слов."<br>
Ответ:<br>
```
Установите аэраторы на краны для снижения расхода воды.
Используйте дождевую воду для полива растений.
Выключайте воду во время чистки зубов.
Проверяйте краны на утечки и ремонтируйте их.
Используйте посудомоечную машину только при полной загрузке.
```
## Formatting (Форматирование)
Когда требуется представить информацию в определённой форме, важно указать это в запросе. Форматирование помогает получить ответы, которые сразу можно использовать, будь то таблицы, списки или текстовые блоки.

Чёткий формат облегчает восприятие информации, её копирование и использование в документах, отчётах или презентациях.

**Как задавать формат?**
 - Укажите желаемую структуру.<br>
Пример: "Напиши преимущества и недостатки использования солнечной энергии в виде таблицы с двумя колонками: 'Преимущества' и 'Недостатки'."

 - Задавайте длину ответа.<br>
Пример: "Напиши описание продукта в одном абзаце до 100 слов."

 - Используйте списки, если требуется чёткая структура.<br>
Пример: "Предложи 5 идей для блога о здоровье в виде маркированного списка."

Пример:Промпт: "Создай таблицу с 3 колонками: 'Проблема', 'Решение', 'Преимущества'. Опиши, как искусственный интеллект может улучшить медицинскую диагностику."

Ответ:<br>
|Проблема|Решение|Преимущества|
|---|---|---|
|Неправильная диагностика|Анализ медицинских данных ИИ|Повышение точности диагностики|
|Долгое ожидание результатов|Автоматизированные алгоритмы|Ускорение обработки данных|
|Высокая стоимость анализов|Оптимизация процессов|Снижение затрат на лечение|

## Audience Perspective (Перспектива аудитории)
Понимание аудитории — ключевой аспект написания эффективных промптов. Разным группам людей требуется информация, поданная в разных форматах, тоне и уровне сложности.

Если текст не адаптирован под целевую аудиторию, он может быть непонятен или неэффективен. Например, научные термины не подойдут для детей, а слишком упрощённые объяснения могут разочаровать профессионалов.

**Как адаптировать запросы под аудиторию?**

 - Определите целевую группу.
Пример: "Объясни, как работает солнечная энергия, для детей 10 лет."

 - Задайте уровень сложности.
Пример: "Напиши текст о квантовой физике в доступной форме для широкой аудитории."

 - Укажите предпочтительный тон.
Пример: "Создай текст для молодёжной аудитории в дружелюбном и мотивирующем стиле."

Пример:<br>
Промпт: "Объясни, что такое нейронные сети, для школьников."<br>
Ответ:<br>
"Нейронные сети — это как мозг компьютера. Они помогают ему учиться на примерах. Например, чтобы научить компьютер узнавать кошек, ему показывают много картинок кошек. Чем больше он учится, тем лучше понимает, что перед ним."<br>

Теперь, когда вы освоили все базовые техники, давайте посмотрим, как они могут быть применены в реальной жизни и работе. Ниже приведены примеры использования моделей для автоматизации задач, обучения, маркетинга и других сфер.

## Автоматизация работы
Написание писем<br>
Пример: "Напиши деловое письмо клиенту с предложением сотрудничества в сфере IT."

Создание отчётов.<br>
Пример: "Составь отчёт о трендах в маркетинге за 2023 год в формате из 3 абзацев."

Генерация идей.<br>
Пример: "Предложи 10 идей для улучшения пользовательского интерфейса мобильного приложения."

Объяснение сложных тем.<br>

Пример: "Напиши 5 вопросов с выбором ответа на тему 'История Древнего Египта'."

Создание контента.<br>

Анализ конкурентов.<br>
