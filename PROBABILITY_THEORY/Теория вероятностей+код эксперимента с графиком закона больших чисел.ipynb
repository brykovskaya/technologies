**Эксперимент** — это повторяемый опыт, который может окончиться элементарными исходами (результатами). 

**«Повторяемый»** в этом определении обозначает опыт, который возможно повторить некоторое количество раз при таких же условиях. 

Название **элементарных исходов** отражает их суть: исход либо случился, либо нет. 

Эксперимент, у которого вероятность каждого исхода одинакова, называется **честным**. 

**Множеством** называется некоторый набор неповторяющихся элементов (дни недели, орёл/решка)

Множество A называется **подмножеством множества B**, если любой элемент из A является элементом ножества B.

Пустым называется множество, в котором нет элементов.<img src='images/00001.png'><br>
 

Множество всех элементарных исходов эксперимента называется **вероятностным пространством**. 

Подмножества вероятностного пространства с некоторым количеством элементарных исходов эксперимента называются **событиями**. Среди событий выделяют две особые группы: **достоверные и невозможные события**.<img src='images/00002.png'><br>

Два события называются **пересекающимися**, если есть хотя бы один исход, который входит в оба события. пример: пересекающиеся события — «выбрал корм с тунцом» и «выбрал один из двух кормов».

**Диаграмма Эйлера-Венна**  
На диаграмме всё вероятностное пространство изображается как прямоугольник. Внутри него есть события — они изображены кругами. Если события пересекаются, то и круги пересекаются. Таким образом, если исходов много, удобнее отобразить их на диаграмме, которая наглядно покажет в том числе, пересекаются ли события (есть ли у них общие исходы) или нет.

События называются **взаимоисключающими**, если множества их исходов не имеют общих элементов.

Взаимоисключающие события тоже можно представить с помощью диаграммы Эйлера-Венна. Если в случае с пересекающимися событиями круги, которые обозначают события, пересекаются, то при изображении взаимоисключающих событий круги пересекаться не будут. При этом речь не идёт о ситуациях, где одно событие — подмножество другого. 

**Закон больших чисел**

Закон больших чисел гласит: чем больше число повторений эксперимента, тем ближе относительная частота события к его теоретической вероятности. Чтобы вычислить относительную частоту события, нужно разделить частоту, с которой оно происходит, на общее количество сделанных повторений эксперимента.

тобы проверить, выполняется ли этот закон, можно воспользоваться Python. Смоделируем с помощью него процесс выбора корма и посмотрим, как будет изменяться относительная частота события «выбрал корм с лососем»:

```python
import matplotlib.pyplot as plt
from random import randint

# задаём параметры размера графика
f = plt.figure()
f.set_figwidth(17)
# задаём линию идеальной вероятности
x_points = [1, 151]
y_points = [0.5, 0.5]
plt.plot(x_points, y_points, linestyle='dashed', 
         label="Теоретическая вероятность")
plt.ylim([0, 1.2])


# список x будет хранить порядковые номера испытаний
x = []
# список y будет хранить доли выбранных лососей от текущего числа испытаний
y = []
# количество выбранных лососей
salmon_count = 0

# проводим 150 испытаний (от 1 до 150)
for i in range(1, 151):
    # добавляем текущий номер испытания в x
    x.append(i)
    # симулируем выбор кота: 0 — корм с тунцом, 1 — корм с лососем
    food = randint(0, 1)
    # если выбрал корм с лососем
    if food == 1:
        # увеличиваем количество лососей на 1
        salmon_count += 1
    # подсчитываем текущую долю лососей
    salmon_part = salmon_count / i
    # добавляем её в список y
    y.append(salmon_part)
    # выводим долю лососей через каждые 15 шагов
    if i % 15 == 0:
        print(f'Шаг {i}, доля выпавших лососей на данный момент {salmon_part}')

# строим график
plt.plot(x, y, '-o', label = "Относительная частота лососей")
plt.legend(loc="upper right")
plt.show()
```
