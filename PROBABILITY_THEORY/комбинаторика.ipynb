## Три вида комбинаций — перестановки, размещения и сочетания

### Перестановки. Число перестановок


Перестановками называют упорядоченные наборы одних и тех же элементов, которые отличаются только порядком. Например, 
{1,2,3},{1,3,2},{3,1,2} — это некоторые из возможных перестановок набора целых чисел от 1 до 3. То есть перестановки содержат все элементы исходного набора, но в разном порядке.

**Число перестановок** — это количество всевозможных перестановок набора из некоторых элементов.

Чтобы найти общее количество комбинаций, нужно перемножить количество возможных вариантов на каждом шаге.

Число перестановок n объектов равно произведению всех целых чисел от 1 до n. Его обозначают P n

Число перестановок n объектов равно n!.


```python
from math import factorial

n = 4
res = factorial(n)
print(f'Четыре рекламных поста можно переставить {res} различными способами.')
```

### Размещения. Число размещений

**Размещения** — это комбинации объектов исходного набора с заданным количеством элементов, которые различаются самими элементами и их расположением.

Размещениями из n по k называют все возможные комбинации объектов исходного набора, содержащие по 
k элементов из n объектов исходного набора и различающиеся порядком элементов. При этом n и k  — это целые числа, а n должно быть больше либо равно k.

пример: Подборки, содержащие два фильма из четырёх с различным порядком, — это размещения из 4 по 
2. 

 Число размещений из n по k равно n⋅(n−1)⋅(n−2)⋅...⋅(n−k+1), где 
n  — это количество объектов исходного набора, а 
k  — количество элементов в каждом размещении. То есть нужно найти произведение всех целых чисел от 

n−k+1 до n.






```python
# найти количество размещений 5 букв из русского алфавита:
# from math import factorial

letters_to_take = 5
letters_total = 33

res = factorial(letters_total) / factorial(letters_total - letters_to_take)
print('Количество размещений 5 букв из русского алфавита:', int(res))
```

**Задание 1**   
Сколько музыкальных плейлистов длиной в 5 песен можно составить из 10 треков? Плейлисты с одинаковыми композициями и разным порядком песен считаются разными.

```python
# from math import factorial

playlist_songs = 5
total_songs = 10

playlist_amount = (factorial(total_songs) / 
                   factorial(total_songs - playlist_songs)
                  )
print('Количество плейлистов длиной в 5 песен из 10 треков:', 
      int(playlist_amount))
```

### Сочетания. Число сочетаний

**Сочетания** — это комбинации объектов исходного набора с заданным количеством элементов, которые различаются только самими элементами. У сочетаний нет порядка элементов — это неупорядоченные комбинации.

 Сочетаниями из n по k называют все возможные комбинации объектов исходного набора, содержащие k элементов из n объектов исходного набора и не различающиеся порядком элементов.
 
**Число сочетаний**  
Кроме случая, когда выбирают ровно один элемент, число сочетаний из n по k всегда меньше, чем число размещений из n по k. Потому что, из размещений как бы вычёркиваются все дубликаты с другим порядком тех же элементов. Количество таких дубликатов равно количеству перестановок длины k.



```python
# количество сочетаний 5 букв из русского алфавита
# from math import factorial
letters_to_take = 5
letters_total = 33

res = (factorial(letters_total) / 
       (factorial(letters_to_take) * 
        factorial(letters_total - letters_to_take)))
print('Количество сочетаний 5 букв из русского алфавита:', int(res))
```

**Задание 1**  
Игрок участвует в телевизионной викторине. Требуется отгадать слово из 10 букв. Игрок может «открыть» любые 3 буквы. Сколько вариантов выбора у него есть?

```python
# from math import factorial

word_length = 10
letters_amount = 3 # укажите количество букв, которые можно узнать

choices = (factorial(word_length) /
          (factorial(letters_amount) *
        factorial(word_length - letters_amount))
          )
print(int(choices))
```

```python
# Сколько существует вариантов расставить цифры 5, 6, 7, 8, 9?
# Варианты расстановки пяти чисел можно посчитать с помощью формулы числа перестановок пяти объектов:
#  =n!=5!=120.
res = factorial(5)
print(res)
```

```python
# Родители привели дочку в зоомагазин, чтобы купить ей двух попугаев. 
# Всего в вольере магазина восемь попугаев. Сколько вариантов покупки существует?
# порядок попугаев не имеет значения, нужно найти число сочетаний из 8 по 2.
parrots = 8
pets = 2
combination = (
    factorial(parrots)/
    (factorial(pets)*
    factorial(parrots - pets)
    )
)
print('Количество сочетаний 2 цветов попугаев из 8:', int(combination))
```

```python
# У мальчика есть девять кубиков с цифрой на каждом от 1 до 9.
# Сколько различных трёхзначных чисел можно из них составить?
# Порядок цифр важен:Значит, нужно найти число размещений из 9 по 3
n = 9
k = 3
placement = factorial(n) / factorial(n-k)
print('различных трёхзначных чисел можно составить:', int(placement))
```

```python
#  есть 5 различных клавиатур и 3 различных мышки. 
# Сколько пар комплектов можно из них собрать
# Используя правило умножения, можно найти количество комплектов
n = 5
m = 3
combination = n * m
print('количество комплектов:', int(combination))
```
