Эксперимент — это повторяемый опыт, который может окончиться элементарными исходами (результатами).
Название элементарных исходов отражает их суть: исход либо случился, либо нет.  
Эксперимент, у которого вероятность каждого исхода одинакова, называется честным.  
элементарными называются не только те исходы, чья вероятность одинакова. 

Множеством называется некоторый набор неповторяющихся элементов.  

Множество A называется подмножеством множества B, если любой элемент из A является элементом множества B.

Пустым называется множество, в котором нет элементов, поэтому можно сказать, что пустое множество является подмножеством любого множества.

Подмножества вероятностного пространства с некоторым количеством элементарных исходов эксперимента называются событиями.

Достоверным называют событие, которое точно произойдёт. Такое событие полностью совпадает с вероятностным пространством.

Невозможным называется событие, которое никогда не произойдёт. То есть исходы невозможного события не входят в вероятностное пространство эксперимента, а значит, множество такого события пустое.

Случайным называют событие, которое может как произойти, так и не произойти. 

Два события называются пересекающимися, если есть хотя бы один исход, который входит в оба события.

События называются взаимоисключающими, если множества их исходов не имеют общих элементов.

Вероятность — это числовая мера того, насколько вероятно наступление события в результате случайного эксперимента. Она вычисляется как отношение количества благоприятных исходов к общему количеству всех возможных исходов. Значение вероятности находится в диапазоне от 0 до 1, так как общее количество исходов всегда не меньше, чем число благоприятных условий:




события называются независимыми, если наступление одного события никак не влияет на вероятность наступления другого.

Поскольку события не влияют друг на друга, вероятность их совместного наступления равна произведению их индивидуальных вероятностей.


В сложных случаях можно пользоваться подсказкой:
 - Независимые события: События не влияют друг на друга.  
Примеры: подбрасывание монеты и бросок кубика, многократное подбрасывание монеты (если результаты не влияют друг на друга).
 - Зависимые события: Результат одного события изменяет условия для следующего.  
Примеры: вытягивание карт из колоды без возврата, выбор объектов из набора, когда объекты не возвращаются.

 пересекающимися называются события, которые могут произойти одновременно в одном и том же эксперименте.
 

Вероятность событияA в эксперименте B — это число p от 0 до 1, которое показывает, какую долю всех возможных исходов составляет событие A. Вероятность события A в эксперименте B — это число от 0 до 1, показывающее, какую часть длины, площади или объёма занимает событие A по сравнению с общей длиной, площадью или объёмом вероятностного пространства B.  
Чтобы найти вероятность события с помощью геометрического подхода, нужно разделить длину, площадь или объём, занимаемые событием, на общую длину, площадь или объём вероятностного пространства.


Закон больших чисел гласит: чем больше раз вы повторяете эксперимент, тем ближе относительная частота события будет к его теоретической вероятности.

```python
# Импортируем matplotlib.pyplot для построения графиков.
import matplotlib.pyplot as plt


# Устанавливаем размер фигуры для лучшей читаемости.
plt.figure(figsize=(10, 6))


# Рисуем пунктирную линию, представляющую теоретическую вероятность 0.5
plt.axhline(y=0.5, color='r', linestyle='--', label='Теоретическая вероятность')


# Подписываем оси и устанавливаем заголовок
plt.xlabel('Количество испытаний')
plt.ylabel('Относительная частота')
plt.title('Демонстрация закона больших чисел')
```

```python
# Теперь смоделируем эксперимент несколько раз и запишем результаты.
# Импортируем choice из модуля random для случайного выбора.
from random import choice

# Возможные варианты корма
foods = ['тунец', 'лосось']

# Списки для хранения номеров испытаний и относительных частот
trials = []
relative_frequencies = []

# Счётчик количества выборов лосося
salmon_count = 0

# Общее количество испытаний
total_trials = 150

print('Испытание \t Относительная частота выбора лосося')
# Запускаем цикл симуляции эксперимента:
# результат выбора функции choice() всегда случайный, то 
# результаты могут изменяться от запуска к запуску программы
# Симуляция эксперимента
for trial in range(1, total_trials + 1):
    # Случайный выбор корма
    selected_food = choice(foods)

    # Проверяем, выбрал ли кот лосося
    if selected_food == 'лосось':
        salmon_count += 1

    # Вычисляем относительную частоту выбора лосося
    relative_frequency = salmon_count / trial

    # Добавляем текущий номер испытания и относительную частоту в списки
    trials.append(trial)
    relative_frequencies.append(relative_frequency)

    # Выводим относительную частоту каждые 15 испытаний
    if trial % 15 == 0:
        print(f'\t {trial} \t {relative_frequency:.5f}') 
```

```python
# Строим график относительной частоты
plt.plot(trials, relative_frequencies, 
         label='Относительная частота выбора лосося')

# Отображаем легенду
plt.legend()

# Показываем график
plt.show() 
```

**Случайная величина** — переменная, которая хранит элементарный исход некоторого случайного эксперимента в виде числа.

На практике случайные величины принято делить на две основные категории: дискретные и непрерывные.

дискретная случайная величина принимает ограниченное или счётное множество значений. 



непрерывные случайные величины могут принимать любые значения из некоторого интервала.


**верно для любой случайной величины: сумма вероятностей всех возможных значений случайной величины всегда равна 1**

 **закон распределения дискретной случайной величины описывают при помощи таблицы.**
 
 Чтобы определить вероятность того, что случайная величина примет значение из заданного интервала, необходимо сложить вероятности всех возможных значений, которые лежат в этом промежутке.


**кумулятивная функция называется Cumulative Distribution Function **
CDF — это функция, которая позволяет узнать вероятность того, что случайная величина примет значение меньшее или равное заданному.


**Математическое ожидание** показывает, к какому значению стремится среднее случайной величины при бесконечном числе повторений. 

**По закону больших чисел**, среднее будет приближаться к математическому ожиданию, поскольку с увеличением числа экспериментов вероятности стремятся к своим истинным значениям. Это объясняет, почему среднее значение стремится к математическому ожиданию, хотя на практике они никогда не совпадут точно.


```python
# Код для моделирования бросков игрального кубика

import random
import matplotlib.pyplot as plt

def roll_dice(num_rolls):
    return [random.randint(1, 6) for _ in range(num_rolls)]

def calculate_running_average(rolls):
    averages = []
    total = 0
    for i, roll in enumerate(rolls, start=1):
        total += roll
        current_average = total / i
        averages.append(current_average)
    return averages

# Задаём количество бросков
num_rolls = 100_000

# Выполняем моделирование
rolls = roll_dice(num_rolls)
averages = calculate_running_average(rolls)

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(
    range(1, num_rolls + 1), averages, label="Текущее среднее значение"
)
plt.axhline(y=3.5, color="r", linestyle="--", label="$E[X]=3.5$")
plt.xlabel("Номер попытки")
plt.ylabel("Среднее значение")
plt.title(f"Среднее значение очков при {num_rolls} попытках броска кубика")
plt.legend()
plt.grid(True)

# Подписываем последнюю точку текущим средним значением
last_x = num_rolls
last_y = averages[-1]
plt.annotate(
    f"{last_y:.4f}",
    xy=(last_x, last_y),
    xytext=(last_x, last_y + 0.2),
    arrowprops=dict(arrowstyle="->", color="gray"),
    fontsize=10,
    color="blue",
)

# Подписываем математическое ожидание
plt.annotate(
    "$E[X]=3.5$",
    xy=(1, 3.5),
    xytext=(1, 3.3),
    arrowprops=dict(arrowstyle="->", color="gray"),
    fontsize=10,
    color="red",
)

plt.show()
```
