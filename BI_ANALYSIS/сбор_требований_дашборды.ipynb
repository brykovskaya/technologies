## Назвать график эффективным можно, если он:
 - точно передаёт данные;
 - очевидно читается;
 - компактный;
 - без лишнего украшательства, которое мешает понять смысл.
 
Создать удачную визуализацию хаотично и без подготовки вряд ли получится. Важно следовать таким принципам:
Прежде чем приступить к визуализации, определяем её цель.
 - Подготавливаем данные в соответствии с целью.
 - Учитываем однозначное восприятие, понятность данных и особенности графика при выборе типа визуализации данных.
 - Для оптимального восприятия информации расшифровываем показатели прямо на схеме: под или рядом со столбцами, на секторе и так далее.

**Определяем цель визуализации**
Первым делом поучимся определять цели каждого изученного типа визуализации.

Вот какие цели можно считать основными:
 - Сравнение. -  строятся на тезисах, которые содержат в себе слова «равно», «больше чем», «меньше чем».
 - Сравнение части с целым. - сравнивать значения, но при этом визуально воспринимать их как отношение доли к целому.
 - Изменение во времени. - обычно фокусируются на изменении показателя во времени.
 - Распределение. - показывает, сколько единиц данных попадает в определённые интервалы.
 - Этапы/процесс/поток. - показывают, как с течением времени меняется характер распределения значений по этапам и стадиям.
 
Разберём каждую из них отдельно.

### макет
аналитику нужно заранее позаботиться о том, чтобы результат его работы совпал с ожиданиями заказчика. Как раз в этом случае пригождается макет, или прототип.
Макет — это черновик будущего дашборда.

Этап прототипирования позволяет:
 - визуализировать пожелания и требования заказчика;
 - определить, как будут выглядеть визуализации;
 - показать логику работы и пользовательские сценарии;
 - схематично разработать будущий интерфейс дашборда.
 
Однако к разработке макета нужно подготовиться. Этому этапу предшествуют важные задачи:
1. Подготовить бриф.
1. Понять и изучить аудиторию.
1. Проработать KPI.
1. Выбрать источники данных.

(2) Какая у дашборда цель? Цель — краеугольный камень всей задачи. В ней кроется ценность отчёта для бизнеса, причина для его разработки. Поэтому аналитик должен оценить, зачем заказчику дашборд, что он хочет узнать. Если исходная задумка заказчика не сработает, то нужно предложить альтернативные варианты, которые помогут достичь цели.

(3) Какие KPI и метрики выделить? Чтобы ответить на этот вопрос, нужно понять, какие факторы сильнее всего влияют на решение задач заказчика. Иными словами, какие показатели и какие их значения станут сигналом о том, что всё хорошо, или о том, что нужно срочно что-то менять. Необходимо выделить от трёх до пяти ключевых метрик, которые наиболее важны для решения бизнес-задачи. Например, если задача — повысить рентабельность, то такими показателями станут общий доход, расходы, рентабельность по сегментам и другие.

(4) Как часто заказчик будет обращаться к дашборду? Ответы могут быть разные: каждый час, раз в день или один раз в месяц. Всё зависит от конкретной задачи.
Ответы на все четыре вопроса можно уложить в шаблон, который поможет систематизировать полученную информацию и понять, как же пользователь будет использовать дашборд:

```Как [сотруднику с ролью X], мне нужно [с определённой периодичностью] видеть [эти показатели], чтобы [принять решение о …].```
