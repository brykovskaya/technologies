# Древовидная и точечная диаграммы
## Точечная диаграмма 
```
Точечная диаграмма — это тип графиков, который используют для того, чтобы изучить связь двух показателей, проследить отношение между ними. Если растёт один — влияет ли это как-то на другой? Прямая или обратная зависимость между признаками? На эти вопросы помогает ответить точечная диаграмма.
В ситуации, когда нужно отследить взаимосвязь между тремя признаками, то информацию из одного поля можно передать размером точки. В таком случае диаграмма станет пузырьковой. Обратите внимание: в DataLens нет отдельного вида чарта для пузырьковых диаграмм, но их можно строить, поменяв настройки в точечных.
Задача. Заказчик дашборда о нарушениях сна попросил добавить туда график, на котором будет показано, какие показатели влияют на качество сна пациентов. Точечная диаграмма для этого идеально подойдёт.
Чтобы её построить в конструкторе чартов, нужно выбрать тип графика «Точечная диаграмма». Затем перетащим показатель Качество сна в область «X». В качестве первой гипотезы можно проверить взаимосвязь между качеством сна и количеством шагов. Для этого перетащим поле Ежедневные шаги в область «Y».

На графике ниже уже можно увидеть заметный тренд. Для большей наглядности узнаем, как много пациентов в каждой группе: перетащим поле User ID в область «Размер точек». Диаграмма превратится в пузырьковую.

Последний штрих — показать на графике разделение пациентов по полу. Для этого в область «Цвета» нужно перенести поле Пол. 

Вот каким получился финальный результат. На графике виден устойчивый тренд: с увеличением количества шагов в сутки растёт и оценка качества сна. Цветовая индикация также подсказывает, что женщины лучше оценивают качество сна и проходят больше шагов, чем мужчины. 

Дополнительно можно внести корректировки в диаграмму, чтобы проследить взаимосвязь между средним количеством шагов и диетическими привычками. Для этого нужно перетащить расчётное поле Среднее кол-во шагов в область «Y», а также добавить на точки данные из поля Диетические привычки, перенеся его в область «Точки».
Так влияние среднего количества шагов на оценку качества сна становится ещё заметнее. А данные на точках позволяют увидеть, что питание тоже влияет на оценку качества сна. 

Небольшой комментарий: проследить взаимосвязь между количеством шагов и качеством сна можно и при помощи линейной диаграммы — на этом графике также видна положительная динамика. Но она не покажет другие нюансы, которые мы смогли увидеть благодаря точечной диаграмме, ведь на её точки можно вывести дополнительную информацию.
```

## Древовидная диаграмма
Заказчик также попросил добавить на дашборд чарт, который покажет, как пациенты сочетают две хорошие привычки — правильное питание и дополнительную физическую нагрузку. Он хочет знать, какие комбинации значений этих признаков распространены сильнее всего и насколько велика каждая группа клиентов в абсолютных значениях. Размеры групп нужно сравнить визуально, сопоставив их друг с другом на одном графике.
Для этой задачи отлично подойдёт древовидная диаграмма:
Она предназначена для анализа структуры выборки и вмещает много сегментов, оставаясь ясной.

Древовидную диаграмму можно настроить так, что при наведении курсора на каждый сегмент пользователь сможет увидеть, сколько пациентов входит в группу.

# Геоаналитика
## Виды геоданных
Геоаналитика предполагает работу с геоданными — различными географическими объектами. Геоданные задаются географическими координатами. Система слегка похожа на ту, что используется для обозначения объектов на координатной плоскости, где расположены фигуры, графики и функции.
Виды геоданных:
 - `Точка` — одиночный объект на карте. Задаётся одной парой координат. Пример: `55.752507, 37.623150` . Первое число — это широта, второе — долгота.
 - `Линия` — несколько точек, которые соединены в определённом порядке. Задаётся координатами для каждой точки и порядком их соединения — направлением. Пример:  ` 55.752507, 37.623150 -> 55.756637, 37.619544  `.
 - `Полилиния` — несколько линий, соединённых в один объект. Примеры: `  55.752507, 37.623150 -> 55.756637, 37.619544 -> 55.760039, 37.620446 -> 55.760190, 37.613049  `. В жизни вы часто встречаетесь с полилиниями: на карте метро или карте пробок.
 - `Полигон` — элемент, который состоит из замкнутой линии и области внутри неё. Пример: ` 55.752507, 37.623150; 55.756637, 37.619544; 55.756806, 37.622727; 55.753460, 37.621601  `. На карте мира территории многих стран можно представить как полигон. Например, Люксембург
 - `Мультиполигон` — полигон сложной формы (например, с отверстием) или несколько полигонов, связанных в один объект. Пример мультиполигона на карте — Италия, у которой к основной территории на Апеннинском полуострове примыкают Сардиния, Сицилия и ряд более мелких островов. А в самой стране есть «дырки»: территории Ватикана и Сан-Марино.

## Типы геоданных в DataLens
В DataLens есть два вида геоданных: `геоточка и геополигон`. 
 - Тип «Геоточка» соответствует описанному выше типу геоданных «Точка». Формат записи координат геоточек — `[55.7912,37.6872]`. Геоточки можно использовать, чтобы добавлять на карту точки и полилинии. 

 - Тип «Геополигон» подходит, чтобы изображать полигоны и мультиполигоны. Формат записи координат для геополигонов: `[[[55.60807, 37.5698], [55.60847, 37.56992], [55.60851, 37.57095]]]`.

## Создание карт в DataLens
В DataLens есть пять видов чартов для геоаналитики: 
 - точки,
 - точки с кластеризацией,
 - полилинии,
 - полигоны,
 - тепловая карта.

В этом разделе вы узнаете, как создавать и настраивать каждый из них.

В качестве источника данных воспользуйтесь подключением, в котором содержатся данные о сезонных консультациях по прививкам от гриппа и заказах на поставку прививок. Вас интересует датасет `consultation_vaccination_geo`.

Описание полей:
 - `id` — номер региона, где проводилась консультация.
 - `region` — регион проведения консультации.
 - `revenue` — выручка от заказа на прививку.
 - `order` — количество заказов на прививку.
 - `visitor` — количество посетителей консультации.
 - `region_order` — порядковый номер региона в рамках сезона консультаций.
 - `consultation_season` — сезон, в котором состоялась консультация.
 - `coord_point` — точка региона.
 - `coords_polygon` — полигон региона.
Чтобы визуализации легко читалась, стоит переименовать поля, которые нужны для карт. Поле `region` назовите Регион, `revenue` — Выручка, `order` — Заказы, `visitor` — Посетители, `consultation_season` — Сезон.

### Практика
В этом уроке вам предстоит на основе описанного выше датасета создать шесть чартов:
1. Точки для суммы выручки.
1. Точки с кластеризацией для выручки и сезона консультации.
1. Полилинии для маршрутов консультаций каждого сезона.
1. Полигоны для суммы выручки.
1. Тепловая карта.
1. Карта с несколькими слоями для объёма выручки и количества посетителей.

Чтобы построить графики, ниже даны инструкции и референсы из урока. Не забывайте про подсказки (тултипы) к картам.

#### Точки
Предположим, аналитику нужно показать сумму выручки, полученную в каждом регионе, и сравнить эти значения. Для этого воспользуйтесь картой типа` «Точки (Геоточки)»`: она полезна в ситуациях, когда на карте есть конкретные места — регионы, города, магазины — и вам нужно вывести метрики этих объектов.

В датасете есть поле `coord_point`, которое содержит координаты центра каждого региона — те самые точки, что нужны для выполнения задачи.

Вот как построить нужную визуализацию:
 - Создайте чарт «Карта».
 - Выберите тип карты «Точки (Геоточки)».
 - Перетащите поле coord_point с координатами геоточек в секцию под меню с выбором типа карты.

На карте появились точки для тех регионов, где проводились консультации. Теперь сделайте так, чтобы размер и цвет точки обозначали объём выручки. Для этого:
 - Перетащите поле Выручка в секцию «Размер точек».
 - Перетащите поле Выручка в секцию «Цвета».

Чем больше точка — тем больше выручка! Размер настроен верно. Цвета же были присвоены каждому региону случайно, это не даёт пользователю никакой информации. Лучше поменять их на градиент с таким правилом: чем больше выручка, тем темнее цвет. 

Чтобы это сделать:
 - Зайдите в настройки секции «Цвет».
 - Выберите тип заливки «Градиент».
 - Выберите цвет, который вам нравится, — например, фиолетовый цвет.
 - Поменяйте направление градиента.

Осталось только добавить подписи к данным, чтобы на каждой точке появилась сумма выручки. Для этого перетащите поле Выручка в секцию «Подписи». Также полезно настроить всплывающие подсказки. Чтобы пользователь при наведении курсора на точку видел данные о регионе из полей Регион, Посетители, Заказы и Выручка, перетащите их в секцию «Тултипы».

#### Точки с кластеризацией
На второй карте дашборда нужно не только показать выручку, но и отдельно подсветить сезон проведения консультации в регионе. Также заказчик попросил сделать так, чтобы на большом отдалении точки не сливались в одну кучу, а компактно группировались.

Этого можно добиться с помощью карты типа «Точки с кластеризацией (Геоточки)». Его используют, когда на карте множество точек, которые нужно объединять в кластер, если они стоят рядом. 

Для этого:
 - Создайте чарт «Карта».
 - Выберите тип чарта «Точки с кластеризацией (Геоточки)».
 - Перетащите поле coord_point с координатами геоточек в секцию под меню с выбором типа карты.

На карте стоящие рядом точки объединяются в кластеры. При нажатии на кластер карта приближается до того уровня, на котором будет видна каждая отдельная точка в кластере.

Как и в случае с картой «Точки», далее нужно перетащить поле Выручка в секцию «Размер точек». Чтобы добавить на карту сезон консультации, перетащите поле Сезон в секцию «Цвета». Перейдите в настройки секции и подберите спокойные цвета для обозначения каждого сезона.

Теперь цвет точки обозначает сезон, в который проводилась консультация в этом регионе. Остаётся добавить «Подписи» и «Тултипы» как на предыдущей карте.

#### Полилинии
Полилинии обычно используют, чтобы изобразить траектории движения: перемещение объекта между рядом последовательных точек. Заказчик попросил вас создать карту с полилиниями, на которой будет видно маршруты консультаций в каждом сезоне.

Для этого:
 - Создайте чарт «Карта».
 - Выберите тип чарта «Полилинии (Геоточки)».
 - Перетащите поле coord_point с координатами геоточек в секцию под меню с выбором типа карты.

На карте появились линии. Теперь укажите, как их группировать и раскрасить. Для этого нужно перетащить поле Сезон в секторы «Группировка» и «Цвета». Ради единообразия выберите для сезонов цвета, аналогичные тем, что настроили в чарте «Точки с кластеризацией».

Теперь регионы, где проходили консультации в рамках одного сезона, соединены линией одного цвета. Но сейчас линия просто соединяет все указанные в поле точки. Чтобы полилиния связала регионы именно в том порядке, в котором их посещали консультанты, нужно перетащить поле region_order в секцию «Порядок».

На полилинии можно отобразить другую информацию: например, отразить выручку регионов на маршруте.
Для этого:
 - Перетащите поле Выручка в секцию «Цвета».
 - Настройте градиент как для карты типа «Точки».

Теперь на карте видны все маршруты: чем темнее цвет отрезка, тем больше сумма выручки, полученная в конечном регионе.

#### Геополигоны
Геополигоны используют для визуализации областей на карте — полигонов и мультиполигонов. С их помощью можно показать выручку по регионам, выделив регион не простой точкой, а его границами и занимаемой территорией.

Вот как это сделать:
 - Создайте чарт «Карта».
 - Выберите тип чарта «Полигоны (Геополигоны)».
 - Перетащите поле coord_polygon с координатами в секцию под меню с выбором типа карты.
 - Перетащите поле Выручка в раздел «Цвета».
 - Настройте градиент так же, как делали это для прошлых карт.
 - Перетащите поля Регион, Посетители, Заказы, Выручка в раздел «Тултипы».
На карте выделены регионы, в которых проводились консультации, а цветом обозначен объём принесённой выручки. Чем темнее цвет, тем больше выручка. К примеру, Липецкая область выделена более тёмным оттенком фиолетового в сравнении с соседями, а значит, консультация по прививкам там принесла заметно больше выручки, чем в граничащих с ней областях. Новосибирская, Томская, Кемеровская области и Алтайский край примерно одного цвета, значит, сумма выручки в них сопоставима.

#### Теплокарта
Тепловые карты нужны, чтобы отобразить скопления точек. Области на тепловой карте окрашиваются градиентом от зелёного цвета к красному: чем больше точек сгруппированы в области, тем ближе к красному будет её цвет. В датасете этого урока точек не так много, но некоторые скопления на ней можно увидеть при большом отдалении.
Вот как это сделать:
 - Создайте чарт «Карта».
 - Выберите тип чарта «Теплокарта (Геоточки)».
 - Перетащите поле coord_point с координатами геоточек в секцию под меню с выбором типа карты.

На получившейся карте заметно скопление точек на Северном Кавказе, а если отдалить её ещё сильнее, то станет заметно скопление в Центральной России.

## Ограничение вывода значений селектора
Задача 1. Вам нужно создать дашборд с данными по нескольким регионам. Заказчика интересуют только Ставропольский край, республика Карелия, Омская область, республика Алтай, республика Калмыкия, Еврейская автономная область, Мурманская область, Астраханская область, Саратовская область, Воронежская область. Сейчас в селекторе «Регион» доступно более 60 пунктов, и вручную выбирать все нужные регионы неудобно. 

Поэтому необходимо ограничить список регионов в селекторе до 10 нужных значений. Чтобы это сделать, фильтрация на уровне датасета или чарта не поможет.

Для начала нужно создать отфильтрованное поле `region_filter`:

```
CASE [Регион]
    WHEN 'Ставропольский край' THEN 'Ставропольский край'
    WHEN 'республика Карелия' THEN 'республика Карелия'
    WHEN 'Омская область' THEN 'Омская область'
    WHEN 'республика Алтай' THEN 'республика Алтай'
    WHEN 'республика Калмыкия' THEN 'республика Калмыкия'
    WHEN 'Еврейская автономная область' THEN 'Еврейская автономная область'
    WHEN 'Мурманская область' THEN 'Мурманская область'
    WHEN 'Астраханская область' THEN 'Астраханская область'
    WHEN 'Саратовская область' THEN 'Саратовская область'
    WHEN 'Воронежская область' THEN 'Воронежская область'    
ELSE null
END
```

Далее добавим новое поле в селектор «Регион». Выбор стал гораздо более скромным — чего и нужно было добиться!

## Работа с параметрами
Задача 2. Пользователю нужно сравнить количество посетителей консультаций по прививкам в одном выбранном периоде с количеством заказов в другом. Первый назван отчётным периодом, второй — периодом сравнения. Чтобы это сделать, нужно использовать параметры.
Сперва объединим датасеты consultation_vaccination_geo и consultation_vaccination_date по полю id, номеру региона. Получим датасет “consultation_vaccination_geo - consultation_vaccination_date”. Обратите внимание, что некоторые поля были переименованы для удобства отображения в чартах:
 - id = номер
 - region = регион
 - c_date = дата
   
На основе этих данных нужно создать в датасете четыре параметра: 
 - report_date_from — дата начала отчётного периода.
 - report_date_to — дата конца отчётного периода.
 - compare_date_from — дата начала периода сравнения.
 - compare_date_to — дата конца периода сравнения
   
Тип параметра во всех случаях — «Дата», а значение по умолчанию — начало года.



```
FDIV_SAFE
Синтаксис
FDIV_SAFE( numerator, denominator [ , fallback_value ] )
        

Описание
Возвращает результат деления numerator на denominator. В случае деления на ноль возвращает fallback_value. Если число fallback_value не указано, возвращает NULL.
```
далее нужно создать чарт и новые поля в нём: 
 - `Выручка за отчетный период` — сумма по полю revenue между датами начала и конца отчётного периода:
```
SUM(
    IF(
    DATE([дата]) BETWEEN [report_date_from] AND [report_date_to],
    [revenue],
    NULL)
)
```  

 - `Выручка за период сравнения` — сумма по полю revenue между датами начала и конца периода сравнения:

```
SUM(
    IF(
    (DATE([дата]) BETWEEN [compare_date_from] AND [compare_date_to]) AND [revenue] IS NOT NULL,
    [revenue],
    NULL)
    )
```

 - `Δ Выручка, руб.` — разница между выручкой за отчётный период и выручкой за период сравнения в абсолютных значениях:
```
[Выручка за отчетный период] - [Выручка за период сравнения]
```

 - `Δ Выручка, % `— разница между выручкой за отчётный период и выручкой за период сравнения в относительных значениях:

```
FDIV_SAFE(([Выручка за отчетный период] - [Выручка за период сравнения]), [Выручка за период сравнения])
 у меня не сработала эта формула
поэтому просто по старинке вот так сделала
([Выручка за отчетный период] - [Выручка за период сравнения]) / [Выручка за период сравнения]
```

**`Как считается динамика? Вчера было 100, сегодня стало 120. (120-100)/100 = 20%
Делим на то, с чем сравниваем`**

Далее необходимо добавить получившиеся поля в чарт «Таблица» и настроить условное форматирование. Нужно сделать так, чтобы разница между выручкой за отчётный период и выручкой за период сравнения, то есть дельта выручки, подсвечивалась красным при отрицательных значениях и зелёным — при положительных:

<img src='images/00039.png' width=1000> <br>
Пора переходить к дашборду и добавить на него селекторы для параметров. Начнём с двух селекторов для отчётного периода — параметры report_date_from и report_date_to. Назовите их «Начало» и «Конец».
<img src='images/00040.png' width=1000> <br>
Аналогичным образом добавим два селектора для периода сравнения — параметры compare_date_from и compare_date_to. Их тоже можно назвать «Начало» и «Конец». Осталось дело за малым: добавить для первой группы селекторов заголовок «Отчетный период», для второй — «Период сравнения». Добавим чарт с таблицей.


<img src='images/00041.png' width=500> <br>

Теперь пользователь может выбрать любые даты в селекторах «Начало» и «Конец», а таблица покажет абсолютную и относительную разницу в выручке за выбранные периоды.

## Фильтр в чарте и на дашборде
Задача 3. Представьте, что заказчику нужно убрать из чарта «Топ регионов по выручке» данные по Москве и Санкт-Петербургу и при этом добавить на дашборд фильтр по региону. Если использовать поле Регион и для фильтрации на чарте, и для селектора на дашборде, то после вывода на дашборд в этом чарте всё равно будут показаны все регионы, включая Москву и Санкт-Петербург, которые должны быть исключены.

Если вы задаёте фильтр по полю в чарте и это же поле используете в селекторе на дашборде, то заданные значения фильтрации в чарте сбрасываются. 

Чтобы это исправить, нужно:
 - Дублировать в чарте поле Регион, создав поле Регион (1).
 - Использовать в чарте в фильтре поле Регион (1). Именно на этом шаге нужно отфильтровать ненужные регионы, включив фильтр по полю в чарте.
 - Использовать на дашборде в селекторе поле Регион.

<img src='images/00042.png' width=1000> <br>

<img src='images/000.png' width=1000> <br>
