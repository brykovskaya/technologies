## Назвать график эффективным можно, если он:
 - точно передаёт данные;
 - очевидно читается;
 - компактный;
 - без лишнего украшательства, которое мешает понять смысл.
 
Создать удачную визуализацию хаотично и без подготовки вряд ли получится. Важно следовать таким принципам:
Прежде чем приступить к визуализации, определяем её цель.
 - Подготавливаем данные в соответствии с целью.
 - Учитываем однозначное восприятие, понятность данных и особенности графика при выборе типа визуализации данных.
 - Для оптимального восприятия информации расшифровываем показатели прямо на схеме: под или рядом со столбцами, на секторе и так далее.

**Определяем цель визуализации**
Первым делом поучимся определять цели каждого изученного типа визуализации.

Вот какие цели можно считать основными:
 - Сравнение. -  строятся на тезисах, которые содержат в себе слова «равно», «больше чем», «меньше чем».
 - Сравнение части с целым. - сравнивать значения, но при этом визуально воспринимать их как отношение доли к целому.
 - Изменение во времени. - обычно фокусируются на изменении показателя во времени.
 - Распределение. - показывает, сколько единиц данных попадает в определённые интервалы.
 - Этапы/процесс/поток. - показывают, как с течением времени меняется характер распределения значений по этапам и стадиям.

Time-to-market — это время от старта разработки продукта до его выхода на рынок. Продукт — аналитическое решение, его time-to-market — время от постановки задачи на разработку инструмента до принятия готового решения заказчиком.

Решение `self-service` — это продукт для самостоятельного использования пользователем. Например, сервис, где пользователи могут самостоятельно делать выгрузки из базы данных или собирать отчёты из готовых датасетов.

`Time-to-market` будем считать так: вычитать из даты закрытия задачи по продукту дату открытия задачи по продукту.

Также мы хотим видеть этот показатель за разные периоды — год, квартал, месяц, чтобы рассмотреть динамику time-to-market и понять, что тормозит процесс, как избавиться от узких мест и уменьшить значение этой метрики. # Основы синтаксиса Markdown и примеры использования
Вот где используют Markdown:
1. Текстовые редакторы и заметки:
     - Typora — мощный Markdown-редактор с функцией предпросмотра.
     - Obsidian — популярный инструмент для создания заметок и организации знаний.
2. Платформы для документооборота:
     - GitHub поддерживает Markdown для README-файлов и документации.
     - GitLab, аналогично GitHub, использует Markdown для проектной документации.
     - Jupyter — среда разработки для записи, передачи и запуска кода

пример оформления текста:

```
## Отчёт по результатам анализа      

### Введение

Это первый абзац. Он содержит общую информацию о результатах анализа.

### Методы

Это второй абзац. Здесь излагаются методы и методики использования.

### Результаты 

Это третий абзац. В этом разделе — результаты и выводы.

**Элементы:**

1. Элемент 1

2. Элемент 2

3. Элемент 3

#| || **Заголовок1** | **Заголовок2** || 

|| Данные 1 | Данные 2 || 

|| Данные 3 | Данные 4 || 

|#


Результаты исследования — 
[Источник](https://www.opendatanetwork.com/dataset/health.data.ny.gov/cnih-y5dw)
```

Придётся чуть привыкнуть, если до этого вы часто использовали быстрые команды вроде `cntr+b/command+b`, вместо неё в Markdown нужно набирать звёздочки: `**text**`. 

Иногда стандартные методы не работают: например, выделять текст цветом в Jupyter нужно с помощью  `$\color{red}{\text{Марс}}$` . Поэтому важно учитывать возможности платформы.

## Списки
1) Простой неупорядоченный список

Он отображает элементы, которые не имеют определённого порядка или приоритета. Элементы здесь выделяем такими маркерами: `-`, `*` или `+`. 

Разметка:
```
* Элемент 1
* Элемент 2
* Элемент 3
```
* Элемент 1
* Элемент 2
* Элемент 3

2) Вложенный неупорядоченный список

Он позволяет создать иерархию элементов в неупорядоченном списке. Здесь добавляем отступы перед маркером вложенного элемента, чтобы указать подсписок. 

```
- Элемент 1
   - Элемент A
   - Элемент B
- Элемент 2
```

- Элемент 1
   - Элемент A
   - Элемент B
- Элемент 2

3) Простой упорядоченный список
```
1. Первый пункт
1. Второй пункт
1. Третий пункт
```
1. Первый пункт
1. Второй пункт
1. Третий пункт.

4) Вложенный упорядоченный список

```
1. Первый пункт
   1. Вложенный пункт
   1. Вложенный пункт
1. Второй пункт
```
1. Первый пункт
   1. Вложенный пункт
   1. Вложенный пункт
1. Второй пункт

5) Чекбоксы 
Они позволяют создавать списки задач или контрольные перечни. Каждый элемент списка дополняем квадратными скобками: `[ ]` — для невыполненной задачи, `[x]` — для выполненной. 

```
- [x] Отмеченный пункт
- [ ] Неотмеченный пункт
```
- [x] Отмеченный пункт
- [ ] Неотмеченный пункт

## Сворачивающийся/разворачивающийся блок (спойлер) 
Этот элемент позволяет скрыть определённый контент и раскрыть его по необходимости. Он сокращает объём отображаемой информации, и документ становится компактнее и удобнее для чтения.
Спойлеры часто используют, чтобы скрыть ответы на вопросы, детали сюжета или любую другую информацию, которую пользователь пожелает прочитать позже. 
В стандартном Markdown нет встроенной поддержки для создания сворачивающихся блоков. Однако многие платформы и расширения, как GitHub и некоторые движки для блогов, имеют свою реализацию этой функции. 
Разметка:
```
<details>
  <summary>Нажмите, чтобы развернуть</summary>
  Здесь ваш скрытый контент или текст, который будет виден, 
  только если пользователь нажмёт на "Нажмите, чтобы развернуть".
</details>

```

Разметка:

Для более длинных или сложных фрагментов используется отдельный блок кода, обрамлённый тройными обратными кавычками. Это выделяет код из основного текста. Так часто демонстрируют примеры сложных SQL-запросов или скриптов.

# Применение Markdown на дашбордах
Работаем с текстовым блоком
Для работы с текстом в DataLens есть несколько способов. Первый способ — использование Markdown. Второй — визуальный редактор, также известный как `WYSIWYG` (англ. what you see is what you get — «что видишь, то и получишь»), но не все платформы поддерживают этот вариант.

Сейчас мы повторим основы форматирования с помощью `Yandex Flavored Markdown` — диалекта Markdown, который основан на CommonMark Spec (спецификация CMS) и используется на многих популярных платформах.

Также посмотрим, как можно редактировать текст с помощью быстрых команд в `WYSIWYG-редакторе`.

По умолчанию редактор текста работает в формате WYSIWYG-редактора для переключения. Выберем в настройках режим «Разметка Markdown»:

## Функции разметки
### Жирный

Такая разметка отлично подходит, если мы хотим выделить текст, но не хотим увеличивать межстрочное расстояние. Давайте выделим некоторые категории, которые не стали заголовками. 
Как используем:
YFM. Здесь снова пригодятся звёзды: ставим ** без пробелов с обеих сторон.
`**Категория пользователей с заболеванием сердца:** `
WYSIWYG. Чтобы активировать, пользуемся кнопкой на панели или быстрой командой `Ctrl + B (⌘ + B)`.

### Курсивный

Давайте пометим курсивом комментарий, который располагается под индикаторами. 
Как используем:
YFM. Добавляем с обеих сторон по одной * без пробелов.
`*При рассмотрении средних показателей лучше смотреть срез по определённому полу и возрастной категории* `
Перед словом и после него ставим по одной звёздочке, без пробелов. В итоге шрифт слова становится курсивным.

### Зачёркнутый

YFM. Лишний или устаревший текст зачёркиваем с помощью ~: без пробелов и с обеих сторон.
`~~зачёркнутый~~ `
WYSIWYG. Чтобы активировать, пользуемся кнопкой на панели или быстрой командой `Ctrl + Shift + S (shift + ⌘ + S)`.

### Подчёркнутый

YFM. Подчёркиваем важный момент в тексте с помощью ++: тоже без пробелов и с обеих сторон.
`++подчёркнутый++ `

WYSIWYG. Чтобы активировать, пользуемся только быстрой командой `Ctrl + U (⌘ + U)`, так как  кнопка на панели отсутствует.

Как используем:

## Спойлер

Спойлер используем, чтобы хранить крупный объём информации, который не всегда нужен пользователю. Или когда хотим скрыть детали, которые пользователь открывает по запросу.
Как используем:

YFM. В прошлом уроке вы уже видели форматирование спойлера, но в DataLens он задаётся иначе. Вот как оформляем сворачивающийся/разворачивающийся блок в DataLens:

`{% cut "ИМТ (индекс массы тела)" %}`

величина, позволяющая оценить степень соответствия массы человека и его роста и тем самым косвенно судить о том, является ли масса недостаточной, нормальной или избыточной.

`{% endcut %} `

Заголовок спойлера задаём с помощью` {% cut "ИМТ (Индекс Массы Тела)" %}`. Дальше идёт текст, который мы скрываем под спойлером и раскрываем после нажатия. В конце не забываем поставить` {% endcut %}`. 

WYSIWYG. Чтобы активировать, пользуемся кнопкой на панели или быстрой командой `Ctrl + Shift + 7 (⌥ + ⌘ + 7)`.

## Текстовые блоки

Вспомним из урока про важность юзабилити и UX, что зрение пользователя чаще всего перемещается по F-паттерну. Поэтому в левом верхнем углу стоит расположить самую важную информацию: текстовый блок с цветовой индикацией отлично подходит на эту позицию. 

Как используем:

YFM. Текстовые блоки открываются — `{% note тип_блока "Заголовок" %}` и закрываются — `{% endnote %}` через упоминание `note`. 

Есть несколько видов текстовых блоков — Тип_блока:

 - `info` — «информация», выделяется синим цветом и сообщает о полезной информации.
 - `tip` — «совет», выделяется зелёным цветом и помогает пользоваться представленной информацией.
 - `warning` — «важно», выделяется оранжевым цветом и подчёркивает важность информации.
 - `alert` — «внимание», выделяется красным цветом, это поле с важной информацией.

WYSIWYG. Чтобы активировать, пользуемся кнопкой на панели или быстрой командой `Ctrl + Shift + 8 (⌥ + ⌘ + 8)`.

## Разделитель

Разделитель можно использовать как внутри самого текста, так и в начале или конце текстового блока. Такой вариант пригодится, если у нас, например, два текстовых блока рядом: один с прокруткой, а второй — без.
Как используем:
YFM. «Возводим границы» с помощью трёх минусов: `−−−`.

Как используем:
YFM. Можем даже использовать несколько цитат одновременно.

WYSIWYG. Чтобы активировать, пользуемся кнопкой на панели или быстрой командой `Ctrl + Shift + . (shift + ⌘ + .)`.

## Эмодзи

## Индикаторы
Главные показатели, которые нужно отслеживать персоналу клиники, — количество пациентов на их попечении и их среднее качество сна. Чтобы их отобразить, достаточно показать только одно число, для этого идеально подойдёт визуализация «Индикатор».

Два показателя легко отобразить на двух индикаторах, но в этой ситуации можно сэкономить пространство на дашборде. Для этого нужно `при помощи дополнительной настройки индикатора уместить на нём сразу два значения`.

Чтобы создать такой индикатор, воспользуйтесь функциями разметки. Они позволяют гибко форматировать текст и настраивать визуализации под нужды заказчика. В этом уроке рассмотрим только основные функции: 
 - `BR()`, 
 - `COLOR()`, 
 - `SIZE()`, 
 - `BOLD()`, 
 - `URL()`. 

Подробнее о функциях разметки для DataLens можно почитать в документации.

Чтобы получить нужную визуализацию, нужно сделать индикатор на основе поля, созданного с помощью такой формулы:

`COLOR(size("Всего пациентов","22px"), "#bc5090")+BR() +
size(str(COUNT([User ID])),"43px")+BR() +
COLOR("среднее качество сна: ", "var(--g-color-text-secondary)") +
BOLD(size(str(round(AVG([Качество сна]))),"12px")) `

В результате получится такой индикатор:

Разберём каждую строку отдельно, чтобы понять, что делает каждая функция и как она устроена.

1. Первая строка: `COLOR(size("Всего пациентов","22px"), "#bc5090")+BR()+`.
 С помощью этой строки создана надпись «Всего пациентов».

Функция `COLOR(text, color)` позволяет указать цвет текста. Принимает два аргумента:
 - `text` — переданный функции текст, который нужно отобразить на чарте. В этом примере принимает значение `size("Всего пациентов","22px")`.
 - `color` — цвет текста. В этом примере принимает значение `"#bc5090"`.
Рекомендуем использовать для указания цветов переменные из палитры дизайн-системы `Gravity UI`. Эти цвета будут хорошо различимы как в светлой, так и в тёмной теме.

Функция `SIZE(text, size)` позволяет настраивать размер текста. Принимает два аргумента:
 - `text` — переданный функции текст. В этом примере принимает значение `"Всего пациентов"`.
 - `size` — размер текста, измеряемый в пикселях. В этом примере принимает значение `"22px"`.

Функция `BR()` осуществляет перенос строки.

Символ` + `обозначает склеивание строк.

2. Вторая строка: `size(str(COUNT([User ID])),"43px")+BR()+` .
С помощью этой строки на индикатор добавлен подсчёт количества пациентов в данных.

`SIZE()` определяет размер текста в 43 пикселя.

`BR()` отвечает за перенос строки.

`Функция COUNT()` подсчитывает количество строк. Её вариация — `COUNTD()`, она подсчитывает только строки с уникальным значением. 

В примере её использование избыточно, поскольку ID каждого пациента уникален и каждому пациенту соответствует только одна строка в датасете. Поэтому для верного результата достаточно передать `COUNT() столбец [User ID]`.

`Функция STR()` преобразует числовые и другие типы значений в строковые. В этом примере нужно её использовать, потому что функции разметки работают только со строковыми значениями, то есть с текстом.

3. Третья строка: `COLOR("среднее качество сна:", "var(--g-color-text-secondary)")`.

С помощью этой строки создана надпись `«среднее качество сна:»`.

Функцией `COLOR()` передан текст `"среднее качество сна:"` и цвет `var(--g-color-text-secondary)"`.

4. Четвёртая строка: ` BOLD(size(str(round(AVG([Качество сна]))),"12px"))` .

С помощью этой строки на индикатор добавлено среднее значение поля `«Качество сна»`.

`SIZE()` определяет размер текста в 12 пикселей.

`STR() ` преобразует число в строку.

`Функция AVG()` подсчитывает среднее значение. Вместо неё можно использовать такую формулу: `SUM([Качество сна])/COUNT([User ID])`. 

В качестве значения функции указываем столбец `Качество сна`.

`Функция ROUND(number[,precision])` округляет значение, указанное в аргументе `number`, до указанного в аргументе `precision` количества знаком после запятой. Если число precision не указано, то number округляется до ближайшего целого.

`Функция BOLD( text )`  выделяет переданный ей текст жирным шрифтом.

Рассмотрим, какие действия нужно сделать, чтобы получить второй вариант форматирования таблицы.

1. Выбрать тип чарта «Таблица».
2. Создать несколько вычисляемых полей. Это можно сделать как в датасете, так и при создании чарта. Выбирайте наиболее подходящий вам способ.
    - Создать вычисляемое поле, в котором среднее количество пройденных за сутки шагов будет помечено цветным маркером, показывающим, много это или мало. За основу для оценки возьмём данные «Международного журнала нутрициологии поведения и физической активности»: менее 5000 шагов в сутки — мало (красный маркер), от 5000 до 8000 — средняя активность (жёлтый маркер), более высокие значения — много (зелёный маркер). Формула для вычисляемого поля:
```
IF (
   avg(int([Ежедневные шаги])) < 5000, str( round( avg(int([Ежедневные шаги])) ) ) + "🔴",
   avg(int([Ежедневные шаги])) < 8000, str( round( avg(int([Ежедневные шаги])) ) ) + "🟠",
   str( round( avg(int([Ежедневные шаги])) ) ) + "🟢"
)
```
   - Создать вычисляемое поле, которое объединит данные столбцов «Нарушение сна» и «Использование лекарств». Жалобы пациента на нарушение сна будет обозначать символ "😵‍💫", приём лекарственных средств для — "💊". Оба варианта можно отметить комбинацией "💊 + 😵‍💫", а отсутствие проблем со сном — маркером "❎". Формула для вычисляемого поля:
```
IF (
   [Нарушения сна] = 1 AND [Использование лекарств] = 1, "💊 + 😵‍💫",
   [Нарушения сна] = 1 AND [Использование лекарств] = 0, "😵‍💫",
   [Нарушения сна] = 0 AND [Использование лекарств] = 1, "💊",
   "❎"
)
```
   - Создать вычисляемое поле для того, чтобы сортировка по времени отхода ко сну работала правильно. Дело в том, что по умолчанию значение 00:00 будет восприниматься DataLens как меньшее в сравнении с 23:45, хотя на самом деле пациент, который пошёл спать в 00:00, лёг позже, чем тот, кто отошёл ко сну в 23:45. Нужно сделать так, чтобы в поле со временем был учтён ночной переход. Для этого воспользуемся небольшой хитростью: создадим поле «Сортировка времени», используя следующую формулу:

```
CASE [Время отхода ко сну]  
   WHEN '22:00' THEN 1  
   WHEN '22:15' THEN 2  
   WHEN '22:30' THEN 3  
   WHEN '22:45' THEN 4  
   WHEN '23:00' THEN 5  
   WHEN '23:15' THEN 6  
   WHEN '23:30' THEN 7  
   WHEN '23:45' THEN 8  
   WHEN '00:00' THEN 9  
   WHEN '00:15' THEN 10  
   WHEN '00:30' THEN 11  
   WHEN '00:45' THEN 12  
   WHEN '01:00' THEN 13  
   WHEN '01:15' THEN 14  
   WHEN '01:30' THEN 15  
   ELSE 16  
END   
```
3. Настроить «Линейный индикатор» с градиентом внутри таблицы. Для этого нужно перейти в настройки поля «Качество сна» и включить опцию «Линейный индикатор», выбрав тип заливки «Градиент». Осталось настроить палитру, выбрав тип «Двухцветный» и задав минимальное значение 0 и максимальное 10.
4. Переместить все нужные поля в соответствующие секции конструктора чартов.

5. Добавление ссылки в поле

Ещё одна полезная функция разметки — `URL()`. Она превращает переданный ей текст в ссылку, которая будет вести по выбранному пользователем URL-адресу. При переходе по ссылке страница откроется в новой вкладке браузера.
URL() принимает два аргумента:
 - `address` — ссылка на страницу, куда нужно отсылать пользователя при нажатии на текст.
 - `text` — текст в гиперссылке, поддерживает разметку.

В качестве текста можно использовать любые символы Unicode, в том числе эмодзи. Поле «Источник: норма шагов» создано формулой ниже:
`URL('https://ijbnpa.biomedcentral.com/articles/10.1186/1479-5868-8-79','🔗') `

Функция разметки URL() позволяет дополнить ссылку данными, хранящимися в других полях. К примеру, если в качестве ссылки указать выдачу поискового запроса по фразе «описание препарата», то можно добавить к запросу название препарата, хранящееся в отдельном столбце данных, чтобы по нажатию на ссылку искать информацию о конкретном лекарстве. В таком случае не нужно будет давать ссылки на каждый препарат, достаточно будет воспользоваться такой формулой:
`URL('https://yandex.ru/search/?text=описание+препарата+' + [название лекарственного препарата], [название лекарственного препарата]) `

## Линейчатая диаграмма
Линейчатая диаграмма отлично подходит, если вам нужно сравнить, как различается какой-то показатель среди нескольких групп внутри выборки. Например, чтобы сравнить, кто из пациентов в среднем совершает больше шагов в день — мужчины или женщины. На линейчатую диаграмму также можно добавить маркеры, которые будут показывать какую-то дополнительную характеристику значения показателя для группы: например, тот факт, что оно выше среднего.

Далее разберём, как добавить к значениям такие подписи:

<img src='images/00015.png' width=400><br>

Для вывода стрелок нужно создать дополнительное поле `[Стрелки]` по такой формуле:
```
IF(
   [Ср. количество шагов] < AVG( [Ср. количество шагов] TOTAL ), STR( ROUND([Ср. количество шагов]) ) + ' ⬇',
   [Ср. количество шагов] > AVG( [Ср. количество шагов] TOTAL ), STR( ROUND([Ср. количество шагов]) ) + ' ⬆',
   STR( ROUND([Ср. количество шагов]) )
)
```
Для сопоставления значений по категории со средним по всей выборке использованы функции AVG и TOTAL. Эта комбинация позволяет подсчитать общее среднее без разбивки на группы. 
На получившемся чарте также присутствуют маркеры, которые показывают оценку количества пройденных в среднем шагов. Чтобы их добавить, создадим новое поле, которое одновременно будет содержать и маркеры, и данные из поля `[Стрелки]`. Вот формула для такого действия:
```
IF (
       AVG(INT([Ежедневные шаги])) < 5000, [Стрелки] + "🔴",
       AVG(INT([Ежедневные шаги])) < 8000, [Стрелки] + "🟠",
       [Стрелки] + "🟢"
   ) 
```

<img src='images/00016.png' width=1000><br>
Похожие диаграммы можно создать и для сравнения других средних показателей: качества сна и сожжённых калорий. Они будут отличаться тем, что на них не будет цветного маркера, только стрелка.

На линейчатой диаграмме, расположенной ниже других, стоит включить опцию «Легенда», чтобы подписать, что обозначают цвета. На этом чарте также нужно включить в настройках подписи «Перекрытие подписей», для того чтобы значения показывались без ограничения по размеру чарта.

## Единицы измерения значений
Если вам нужно подписать на графике единицы измерения — `«%», «шт.», «млн.»`, это также можно сделать в формуле.
Например, чтобы вывести количество шагов в тысячах и подписать это обозначение, можно использовать такую формулу:
`CONCAT("🔴 ", STR(ROUND([Ежедневные шаги]/1000,2)), "тыс.") `

Логика обозначений: если полученное значение больше или равно 100%, рядом с ним должен стоять зелёный круг, если меньше 100% — красный кружок.

Выберите правильную формулу, чтобы выполнить эту задачу.
```
IF(
    [complete_plan] >= 1, CONCAT("🟢 ", STR(ROUND([complete_plan]*100,1)), "%"),
    IF(
        [complete_plan] < 1, CONCAT("🔴 ", STR(ROUND([complete_plan]*100,1)), "%"),
        CONCAT(STR(ROUND([complete_plan]*100,1)), "%")
    )
)
```
упрощаем формулу

```
IF ([complete_plan]>=1,
CONCAT ("🟢 ", STR(ROUND([complete_plan]*100,1)), "%"),
CONCAT ("🔴 ", STR(ROUND([complete_plan]*100,1)), "%"))
```

## Сортировка
способы отсортировать значения в чарте в DataLens:
 - `Сортировка по умолчанию` — блок «Сортировка» в чарте. В чартах можно настроить сортировку по разным полям.
 - `Сортировка через дополнительное расчётное поле`. Сортировку можно задать через отдельное поле, указав в нём порядок отсортированных значений.
 - `Сортировка через оконную функцию`.С помощью оконной функции можно отсортировать данные на чарте.

Заказчик передал пожелание: он хочет видеть количество заказов за каждый сезон консультаций в виде рейтинга — от самого обильного на заказы сезона к менее насыщенным. Чтобы выполнить задачу, нужна сортировка. 

Для начала нужно создать линейчатую диаграмму, на которой по оси X будет отложено количество заказов, а по оси Y — название сезона.

 - Москве, 
 - Санкт-Петербурге, 
 - Севастополе, 
 - Ставропольском крае, 
 - республике Карелия, 
 - Мурманской области

Причём порядок должен быть именно таким. Выделить эти регионы среди остальных можно фильтром, однако отсортировать их в нужном порядке не получится. Сортировка возможна или в алфавитном порядке, или по количеству посетителей.

В этой ситуации поможет сортировка с помощью дополнительного расчётного поля. Чтобы расположить регионы в нужном заказчику порядке, создадим поле `region_sort` по формуле:

```
CASE [Регион]
    WHEN "Москва" THEN "1. Москва"
    WHEN "Санкт-Петербург" THEN "2. Санкт-Петербург"
    WHEN "Севастополь" THEN "3. Севастополь"
    WHEN "Ставропольский край" THEN "4. Ставропольский край"    
    WHEN "республика Карелия" THEN "5. республика Карелия"  
    WHEN "Мурманская область" THEN "6. Мурманская область"    
    ELSE null
END
```
В этой формуле использована конструкция CASE для добавления названия региона с порядковым номером. После ELSE нужно применить NULL, так как другие регионы на графике не нужны.

Осталось перенести поле region_sort на ось Y. Теперь чарт отсортирован в том порядке, который хотел заказчик.

## Сортировка с помощью оконных функций
Ещё одна важная возможность, которая есть в DataLens, — сортировка с помощью оконных функций. Она пригодится и для работы с заказчиком: он просит упорядочить строки в таблице консультаций в зависимости от выручки.  

Для этого можно использовать оконную функцию RANK — `RANK(SUM([revenue]))` 

В результате каждой строке будет присвоен свой порядковый номер: строке с наибольшей выручкой — `1`, строке с наименьшей выручкой — `4`.

Если нужно отсортировать регионы по объёму выручки в пределах отдельных сезонов консультаций, это можно сделать с помощью функции `RANK(SUM([revenue]) WITHIN [Сезон консультации])`. Тогда для каждого сезона ранжирование будет проводиться отдельно.

## Области закрепления и сворачивания
Для размещения элементов на дашборде полезно использовать области закрепления и сворачивания. 

Первая область помогает закрепить нужные элементы. Например, там можно зафиксировать фильтры, если дашборд целиком не помещается на одном экране и его нужно листать вниз.

При этом область закрепления ограничена по высоте: она не может быть больше стандартной высоты селектора. Поэтому чарты разместить в этой области не получится. В сворачиваемой же области можно разместить объекты любого размера. Управлять этой областью можно с помощью значка со стрелками в правом углу.

примеры:

Область закрепления
 - «Нужно сделать так, чтобы в любом месте лонгрида было видно, какой сейчас статус по выполнению плана»
 - «Нужно сделать так, чтобы наверху страницы дашборда всегда было видно, какую дату я выбрал в календаре»

Область сворачивания
 - «Необходимо дать пользователю возможность смотреть на эти два чарта, только если они ему интересны. Если они ему не нужны, то на дашборде их быть не должно»
 - «Нужно добавить на дашборд описание расчёта показателей, но так, чтобы его всегда можно было убрать и оно не занимало много места»

## Точечная диаграмма

**Точечная диаграмма** — это тип графиков, который используют для того, чтобы `изучить связь двух показателей`, `проследить отношение между ними`. Если растёт один — влияет ли это как-то на другой? Прямая или обратная зависимость между признаками? На эти вопросы помогает ответить точечная диаграмма.


В ситуации, когда нужно отследить взаимосвязь между тремя признаками, то информацию из одного поля можно передать размером точки. В таком случае диаграмма `станет пузырьковой`. Обратите внимание: в DataLens нет отдельного вида чарта для пузырьковых диаграмм, но их можно строить, поменяв настройки в точечных.

### Задача. 
Заказчик дашборда о нарушениях сна попросил добавить туда график, на котором будет показано, какие показатели влияют на качество сна пациентов. Точечная диаграмма для этого идеально подойдёт.

Чтобы её построить в конструкторе чартов, нужно выбрать тип графика «Точечная диаграмма». Затем перетащим показатель Качество сна в область «X». В качестве первой гипотезы можно проверить взаимосвязь между качеством сна и количеством шагов. Для этого перетащим поле Ежедневные шаги в область «Y».
