**Этапы разработки дашборда**


**Точечная диаграмма** — это график, на котором каждая точка соответствует определённому значению двух переменных. Такую диаграмму используют, чтобы найти зависимость между измерениями и показателями или показать разброс значений.

**Пузырьковая диаграмма** — более сложная версия точечной диаграммы, где появляется третья переменная в виде размера точек. Такая диаграмма позволяет не только увидеть взаимосвязь между двумя переменными, но и оценить, как третья переменная влияет на эту взаимосвязь.

Вот какие задачи решают точечная и пузырьковая диаграммы:
 - Визуализация сложных взаимосвязей. Позволяют наглядно и понятно представить сложные взаимосвязи между несколькими переменными.
 - Анализ тенденций и закономерностей. Помогают выявить тенденции и закономерности в данных, что пригождается при прогнозировании будущих событий или принятии решений.
 - Сравнение данных. Позволяют сравнивать данные между различными группами или периодами времени и находить различия и сходства.
 - Оценка влияния факторов. Пузырьковая диаграмма даёт увидеть, как изменение одного фактора влияет на другие переменные.
 
**Диаграмма воронки** — это тип таблицы или диаграммы, который часто используют, чтобы показать этапы продаж и сумму потенциального дохода для каждого этапа. Такая диаграмма пригождается и когда нужно найти потенциально проблемные области в организации продаж или других последовательных процессах. Диаграмма **«плоское дерево»** — один из способов показать части целого. Древовидная диаграмма выглядит как большой прямоугольник (общее целое), который состоит из разных областей (части целого) разного размера.

диаграммамы:
 - **Каскадные** (с последовательностью и зависимостью задач).ее еще называют waterfall-диаграммой, мостовой или каскадной. С её помощью проводят факторный анализ.
 - **Санкей** (с причинно-следственными связями и влиянием факторов друг на друга).
 - **Ганта** (для планирования проектов и отслеживания прогресса).
 - **Калибровочные**(с диапазонами значений и их распределением).

**Факторный анализ** — это метод, который помогает оценить, как различные факторы влияют на итоговый результат, а каждый элемент — на такие ключевые показатели, как доход, стоимость или объём продаж. Элементы способны одновременно влиять на несколько показателей. Например, от изменения цены на товар зависят его продажи, доход и стоимость запасов.


каскадная диаграмма. пример

Чтобы построить каскадную диаграмму:
Понадобятся определённые исходные данные: начальная точка, факторы, влияющие на изменения, и конечная точка.
Все переменные должны быть в одном измерении: например, нельзя одновременно использовать проценты и абсолютные значения (штуки или рубли).


**Диаграмму Ганта** часто используют для планирования задач: как в крупных проектах, так и в управлении личными делами. Она выглядит как временная шкала выполнения задач и помогает организовывать работу, отслеживать прогресс.
