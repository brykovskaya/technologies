## Проект спринта 13, часть 1

вам предстоит проанализировать требования заказчика, задокументировать их и создать дашборд в соответствии с ними.

Проект состоит из трёх частей. Вот что вам предстоит выполнить в каждой из них:
 - Узнать исходный запрос заказчика, сформулировать вопросы для интервью с ним, провести интервью и задокументировать все требования.
 - Разработать дашборд с учётом исходных требований и информации, полученной на интервью.

### Документирование бизнес-требований

В рамках вашей стажировки в «ЛайфЛайн» поступила новая задача: 

**`вам нужно разработать дашборд для данных о продажах медицинских пробирок самого разного предназначения — для анализа крови, для хранения мазков ПЦР и множества других.`**

Директор отдела продаж просит вас собрать дашборд, который позволит ему следить за метриками своего отдела. Этот же дашборд нужен для оперативной работы менеджеров по работе с поставщиками и менеджеров по продажам.

Краткое описание бизнес-требований заказчика:


 - Нужно, чтобы на первом листе дашборда отображались основные показатели отдела продаж: сумма продаж, количество проданных товаров и средний чек. На нём же должен быть обзорный лист с динамикой показателей, рейтингом товаров и географией продаж.

 - На втором листе расположите информацию для менеджеров. Им нужно видеть сумму продаж за разные периоды времени. Также менеджеры будут отслеживать критичные изменения по поставщикам, типам клиентов и каналам продаж. Добавьте на этот лист фильтры по датам, регионам и поставщикам.


Исходные требования сформулированы в очень общем виде. Вам предстоит дополнить их.

Для этого вам нужно провести интервью с заказчиком и задать ему вопросы. Интервью с заказчиком пройдёт на предстоящем вебинаре. встреча была, требования собрали.

При составлении вопросов также ориентируйтесь на описание данных, с которыми вам придётся работать. Вот какие поля содержат нужные таблицы в базе данных:

 - `date_of_sale` — дата продажи, фиксирует момент совершения сделки.
 - `product_id` — уникальный идентификатор продукта.
 - `product_name` — название продукта.
 - `quantity` — объём или количество проданного продукта.
 - `sale_price` — цена, по которой продукт был реализован.
 - `total_sale_amount` — общая сумма продаж.
 - `region_id` — идентификатор региона, в котором была совершена продажа.
 - `region_name` — название региона.
 - `polygons` — координаты границ региона.
 - `population` — информация о численности населения региона.
 - `supplier_id` — уникальный идентификатор поставщика.
 - `supplier_name` — название поставщика.
 - `customer_type_id` — идентификатор типа покупателя.
 - `customer_type` — тип покупателя (например, розничный или оптовый).
 - `sales_channel_id` — идентификатор канала продажи.
 - `sales_channel` — канал продажи.
 - `category` — категория продукта.

После интервью задокументируйте требования заказчика к разработке дашборда. 

Опирайтесь на изложенную выше информацию и полученные на вебинаре ответы на вопросы. Если нужно освежить знания о том, как собирать требования заказчика, обратитесь к уроку «Анализ и документирование собранной информации» в теме «Проведение интервью с заказчиком для сбора требований». ### Разработка дашборда
Чтобы создать дашборд в DataLens по бизнес-требованиям заказчика, нужно выполнить следующие шаги.

#### Вкладка «Обзор продаж»
1. Создайте панель фильтров на вкладке. Используйте все фильтры, которые запросил заказчик для этой вкладки. Обратите внимание на настройки фильтров: следите за тем, чтобы заказчику было удобно ими пользоваться.
1. Добавьте индикаторы с общими показателями отдела продаж. Визуализируйте основные метрики так, чтобы было понятно, что это за метрика и в чём она измеряется. Нужно показать сумму продаж, количество проданных товаров, среднюю стоимость товара и другие метрики, если необходимо.
1. Покажите динамику показателей с помощью чартов. Постройте отдельные визуализации, на которых будет видна динамика общей суммы продаж и средней стоимости товара. Используйте подходящий вид чартов, сделайте их читаемыми. Добавьте на каждый график возможность менять детализацию данных, чтобы пользователи могли переключаться между динамикой по месяцам и динамикой по кварталам.
1. Составьте рейтинг продуктов. Создайте чарт, на котором будет видно название продукта и количество проданных единиц этого продукта. Сделайте так, чтобы продукты были отсортированы от самых продаваемых к наименее продаваемым. Выделите цветом разные категории продуктов.
1. Покажите взаимосвязь между объёмом и суммой продаж для каждого продукта. На графике нужно показать корреляцию между количеством проданных единиц каждого продукта и суммой от их продажи. Укажите, к какой категории относится каждый продукт. Добавьте в подпись информацию о среднем чеке по этому продукту.
1. Наглядно визуализируйте структуру продаж поставщиков. На одном чарте должно быть видно, какую часть от всех сделок каждого поставщика составили продажи клиентам разных типов. Чарт должен показывать именно вложенную структуру, то есть то, как из отдельных частей (продаж конкретным типам клиентов) складывается целое (общие продажи всех поставщиков).
1. Создайте карту, чтобы отобразить географию суммы продаж. Постройте чарт, который позволит увидеть все регионы, в которых покупали пробирки, и сравнить регионы по общей сумме продаж.

#### Вкладка «Детализация продаж»
1. Создайте панель фильтров на вкладке. Добавьте основные фильтры, которые запросил заказчик. Кроме того, создайте фильтры, которые позволят выбирать базовый период и период для сравнения метрик. Пользователь будет указывать границы периодов.
1. Визуализируйте рейтинг категорий и продуктов по сумме продаж.
    - Постройте график, на котором будет видно, сколько денег принесли продажи: а) разных категорий продуктов; б) разных продуктов. На чарте нужно добавить возможность переключаться между детализацией по категориям и детализацией по продуктам. Отсортируйте столбцы по убыванию.
    - Покажите изменение суммы продаж на графике от базового периода к периоду сравнения. Воспользуйтесь условным форматированием, чтобы выделить положительные и отрицательные изменения с помощью дополнительных символов в подписях.
1. Визуализируйте рейтинг категорий и продуктов по количеству проданных единиц товара. Настройте график аналогично предыдущему.
1. Соберите таблицу. В ней нужно указать поставщиков, каналы продаж и типы клиентов. Для каждого кейса выведите сумму продаж за базовый период и за период сравнения. Покажите разницу в рублях и в процентах. Воспользуйтесь условным форматированием, чтобы выделить положительные и отрицательные изменения с помощью дополнительных символов в ячейках.
#### Вкладка «Аннотация дашборда»
1. Опишите основные метрики на дашборде. Подготовьте описание дашборда и его основных метрик. Для оформления используйте средства Markdown: списки, выделение, разделители, форматирование текста. Описание должно быть написано так, чтобы заказчик мог легко понять, о чём идёт речь.
1. Сделайте краткую аналитическую записку. Проанализируйте данные на дашборде. Используя Markdown, подготовьте небольшую аналитическую записку для пользователей дашборда.
   
Как только вы создадите дашборд, можно будет перейти к последнему шагу — сдаче проекта на проверку.
