üí° **GIN: Generalized Inverted Search Tree** (–û–±–æ–±—â—ë–Ω–Ω–æ–µ –∏–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –¥–µ—Ä–µ–≤–æ –ø–æ–∏—Å–∫–∞) –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø–æ –ø–æ–ª—è–º —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö Greenplum Database Full Text Search –∏–ª–∏ –ø–æ–ª—è–º —Ç–∏–ø–∞ jsonb.
–ü—Ä–∏–Ω—Ü–∏–ø –¥–µ–π—Å—Ç–≤–∏—è –∏–Ω–¥–µ–∫—Å–∞ GIN –ø–æ—Ö–æ–∂ –Ω–∞ Bitmap index, –Ω–æ GIN —Ä–∞–±–æ—Ç–∞–µ—Ç —Å –¥—Ä—É–≥–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö. **–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è GIN-–∏–Ω–¥–µ–∫—Å–∞**
–°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Å –ø–æ–ª–µ–º —Ç–∏–ø–∞ jsonb –∏ –∏–Ω–¥–µ–∫—Å–æ–º GIN:  
```
CREATE TABLE testgin (
  id bigserial,
  DATA jsonb
);
INSERT INTO testgin(DATA) VALUES ('{"name": "Alex"}');
INSERT INTO testgin(DATA) VALUES ('{"name": "Peter"}');
INSERT INTO testgin(DATA) VALUES ('{"surname": "Volkov", "name": "Ivan"}');
CREATE INDEX ON testgin USING gin(DATA); 
```

```
CREATE FUNCTION myfunc_plpgsql (show_id BIGINT)
    RETURNS TABLE (id BIGINT,
                   order_date DATE,
                   country TEXT,
                   sum_usd money)
AS 
$$

DECLARE
    num_rows BIGINT := 0;
BEGIN
    SELECT COUNT(o.id) FROM orders o WHERE o.id=show_id INTO num_rows;
    IF num_rows = 0 THEN RAISE NOTICE 'Rows with specified ID=% not found',show_id;
                    ELSE RETURN QUERY SELECT o2.id, order_date, country, sum_usd FROM orders o2 WHERE o2.id=show_id;
    END IF;
END
$$
LANGUAGE PLPGSQL
;
```

```
–ö–æ–¥-–ø–æ–¥—Å–∫–∞–∑–∫–∞

DROP TABLE IF EXISTS table2;

CREATE TABLE table2 (id int, dt date)
DISTRIBUTED BY (id)
PARTITION BY RANGE (dt) 
    (START ('2023-06-01'::date)
    END ('2023-07-01'::date)
    EVERY (INTERVAL '1 day'));
-- –µ–∂–µ–¥–Ω–µ–≤–Ω—ã–µ –ø–∞—Ä—Ç–∏—Ü–∏–∏ –Ω–∞ –æ–¥–∏–Ω –º–µ—Å—è—Ü

INSERT INTO table2
SELECT 
  id, 
  '2023-06-01'::date + interval '1 day' * (id%30) 
FROM generate_series(1,100000) as id;

ANALYZE table2;
```

–°–æ–∑–¥–∞–π—Ç–µ —Ç–∞–±–ª–∏—Ü—É-–∏–∑–º–µ—Ä–µ–Ω–∏–µ —Å–æ —Å–ø–∏—Å–∫–æ–º –¥–∞—Ç.

```
–ö–æ–¥-–ø–æ–¥—Å–∫–∞–∑–∫–∞

DROP TABLE IF EXISTS date_dim;

CREATE TABLE date_dim AS
SELECT generate_series(1,30) id,
generate_series(TIMESTAMP '2023-06-01', TIMESTAMP '2023-06-30', INTERVAL '1 day')::date AS dt
DISTRIBUTED BY (dt);

ANALYZE date_dim;
```

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø–ª–∞–Ω –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–ø—Ä–æ—Å–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫–æ–≤ Postgres –∏ GPORCA. –ö–∞–∫—É—é —Ä–∞–∑–Ω–∏—Ü—É –≤—ã —É–≤–∏–¥–µ–ª–∏?

–ö–æ–¥-–ø–æ–¥—Å–∫–∞–∑–∫–∞

```
--Postgres –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫
SET optimizer=off;
EXPLAIN ANALYZE SELECT t.*
FROM table2 t 
    INNER JOIN date_dim d ON t.dt=d.dt 
WHERE d.id BETWEEN 15 AND 20;

--GPORCA –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫
SET optimizer=on;
EXPLAIN ANALYZE SELECT t.*
FROM table2 t 
    INNER JOIN date_dim d ON t.dt=d.dt 
WHERE d.id BETWEEN 15 AND 20;
```
