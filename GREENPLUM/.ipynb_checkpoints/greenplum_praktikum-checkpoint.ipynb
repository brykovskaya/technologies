üí° **GIN: Generalized Inverted Search Tree** (–û–±–æ–±—â—ë–Ω–Ω–æ–µ –∏–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –¥–µ—Ä–µ–≤–æ –ø–æ–∏—Å–∫–∞) –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø–æ –ø–æ–ª—è–º —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö Greenplum Database Full Text Search –∏–ª–∏ –ø–æ–ª—è–º —Ç–∏–ø–∞ jsonb.
–ü—Ä–∏–Ω—Ü–∏–ø –¥–µ–π—Å—Ç–≤–∏—è –∏–Ω–¥–µ–∫—Å–∞ GIN –ø–æ—Ö–æ–∂ –Ω–∞ Bitmap index, –Ω–æ GIN —Ä–∞–±–æ—Ç–∞–µ—Ç —Å –¥—Ä—É–≥–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö. **–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è GIN-–∏–Ω–¥–µ–∫—Å–∞**
–°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Å –ø–æ–ª–µ–º —Ç–∏–ø–∞ jsonb –∏ –∏–Ω–¥–µ–∫—Å–æ–º GIN:  
```
CREATE TABLE testgin (
  id bigserial,
  DATA jsonb
);
INSERT INTO testgin(DATA) VALUES ('{"name": "Alex"}');
INSERT INTO testgin(DATA) VALUES ('{"name": "Peter"}');
INSERT INTO testgin(DATA) VALUES ('{"surname": "Volkov", "name": "Ivan"}');
CREATE INDEX ON testgin USING gin(DATA); 
```

```
CREATE FUNCTION myfunc_plpgsql (show_id BIGINT)
    RETURNS TABLE (id BIGINT,
                   order_date DATE,
                   country TEXT,
                   sum_usd money)
AS 
$$

DECLARE
    num_rows BIGINT := 0;
BEGIN
    SELECT COUNT(o.id) FROM orders o WHERE o.id=show_id INTO num_rows;
    IF num_rows = 0 THEN RAISE NOTICE 'Rows with specified ID=% not found',show_id;
                    ELSE RETURN QUERY SELECT o2.id, order_date, country, sum_usd FROM orders o2 WHERE o2.id=show_id;
    END IF;
END
$$
LANGUAGE PLPGSQL
;
```
