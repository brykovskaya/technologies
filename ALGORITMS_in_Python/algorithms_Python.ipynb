### Стек
*Стек* - это линейная структура данных для хранения одномерного списка. Элементы в стеке могут обрабатываться по принципу LIFO(Last-In, First-Out) FIFO(First-In, First-Out). Порядок добавления и удаления элементов определяет характер стека. Новые элементы могут добавляться и удаляться только с одного конца списка.
Операции со стеком:
 - isEmpty - возвращает true если стек пуст
 - push - добавляет новый элемент
 - pop - возвращает элемент, добавленный последним и удаляет его

Создадим класс с именем Stack, в котором опишем все операции, связанные с классом stack.
Данный код реализует стек (stack) — структуру данных, работающую по принципу LIFO (Last In, First Out — «последним пришёл, первым ушёл»).

Основные компоненты кода:
 - Класс Stack — определяет структуру стека
    - Содержит методы для работы со стеком
    - Хранит данные в списке items
 - Метод __init__ — конструктор класса
    - Инициализирует пустой список items при создании объекта
 - Метод isEmpty
    -Проверяет, пуст ли стек
    - Возвращает True, если список пуст, иначе False
- Метод push
    - Добавляет элемент в конец стека
    - Использует метод append() списка
 - Метод pop
    - Удаляет и возвращает последний элемент стека
    - Использует метод pop() списка
 - Метод peek
    - Возвращает последний элемент стека без удаления
    - Получает доступ к элементу по индексу len(self.items)-1
 - Метод size
    - Возвращает количество элементов в стеке
    - Использует функцию len()

````python
class Stack:
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[len(self.items)-1]
    def size(self):
        return(len(self.items))   
````

````python
# создаем стек с четырьмя элементами
stack = Stack()
stack.push('Red')
stack.push('Green')
stack.push('Blue')
stack.push('Yellow')
print('создали стек с четырьмя элементами:')
stack
````

### Очередь
очередь хранит n элементов в одномерной структуре. Элементы добавляются и удаляются по принципу FIFO. Каждая очередь имеет начало и конец. Когда элементы удаляются из начала, операция называетя "удаление из очереди" - `dequeue`. Когда элементы добавляются в конец, операция называется "постановкой в очередь" - `enqueue`

````python
class Queue(object):
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def enqueue(self, item):
        self.items.insert(0, item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items)
````

Данный код реализует очередь (queue) — структуру данных, работающую по принципу FIFO (First In, First Out — «первым пришёл, первым ушёл»).

Основные компоненты кода:
1. Класс Queue
 - Наследует от базового класса object
 - Содержит методы для работы с очередью
 - Хранит элементы в списке items
2. Метод __init__
 - Конструктор класса
 - Инициализирует пустой список items
3. Метод isEmpty
 - Проверяет, пуста ли очередь
 - Сравнивает список с пустым списком
 - Возвращает True или False
4. Метод enqueue
 - Добавляет элемент в начало очереди
 - Использует метод insert(0, item)
 - Новый элемент помещается в позицию с индексом 0
5. Метод dequeue
 - Удаляет и возвращает последний добавленный элемент
 - Использует метод pop()
 - Извлекает элемент из конца списка
6. Метод size
 - Возвращает количество элементов в очереди
 - Использует встроенную функцию len()


````python
queue1 = Queue() # создание очереди
queue1.enqueue(1) # добавление элемента в начало
queue1.enqueue(2) # добавление элемента в начало
print(queue1.dequeue()) # вывод: 1 (первый добавленный элемент)
print(queue1.size()) # вывод: 1
````

Важно отметить, что реализация очереди через список не самая эффективная, так как вставка в начало списка требует сдвига всех элементов. Для более эффективной реализации можно использовать collections.deque

````python
queue = Queue()
queue.enqueue('Red')
queue.enqueue('Green')
queue.enqueue('Blue')
queue.enqueue('Yellow')
display(queue.size())
display(queue.dequeue())
display(queue.dequeue())
````

### Сортировка пузырьком
Сортировка пузырьком bubble sort - самый простой и медленный алгоритм сортировки. Он спроектирован так, что наибольшее значение перемещается вправо по списку на каждой итерации цикла. При наихудшем сценарии производительность этого алгоритма равна О(n**2). Поэтому его стоит испорльзовать только для небольших наборов данных.

В основе сортировки пузырьком лежит ряд итераций, называемых проходами (passes). Для списка размера N нужно совершить N - 1  проходов. Цель первого прохода - вывести наибольшее значение в конец списка. По мере выполнения алгоритма оно будет постепенно перемещаться вправо по списку. 

В процессе сортировки значения соседних элементов сравниваются между собой попарно. Если в паре большее значение находится слева, происходит перестановка (обмен). Это продолжается до тех пор, пока мы не дойдем до конца списка. Затем начинается следующий проход. Его цель переместить следующее по величине значение на предпоследнюю позицию в списке. Последний элемент не затрагивается, поскольку уже был перемещен в нужную позицию на первой итерации.

Проходы выполняются до тех пор, пока все элементы данных не будут расположены в порядке возрастания

````python
# первый проход сортировки пузырьком. код
list = [25,21,22,24,23,27,26]
lastElementIndex = len(list) - 1
print('первый проход сортировки пузырьком:')
print(0, list)

for idx in range(lastElementIndex):
    if list[idx] >list[idx + 1]:
        list[idx], list[idx+1] = list[idx+1], list[idx]
        print(idx+1,list)
````

````python
# реализация сортировки пузырьком
def BubbleSort(list):
    lastElementIndex = len(list) - 1
    for passNo in range(lastElementIndex, 0, -1): # организует проходы по списку/ # получение последовательности чисел в обратном порядке 
        for idx in range (passNo):
            if list[idx]>list[idx+1]:
                list[idx], list[idx+1] = list[idx+1], list[idx]
    return list
````

````python
list1 = [25,21,22,24,23,127,26]
print('ДО сортировки пузырьком:')
print(list1)
BubbleSort(list1)
print('После сортировки пузырьком:')
print(list1)
````

Анализ производительности сортировки пузырьком.

Сортировка пузырьком включает в себя два уровня циклов:
 - внешний цикл. Совокупность проходов. Например, первый проход - это первая итерация внешнего цикла.
 - внутренний цикл. оставшиеся элементы в списке сортируются до тех пор, пока наибольшее значение не окажется справа. На первом проходе будет N-1 сравнений, на втором N -2. На каждом последующем проходе количество сравнений будет уменьшаться на единицу.

Из-за двух уровней цикличности наихудшая сложность алгоритма равна О(n**2)

### Сортировка вставками
Основная идея сортировки вставками заключается в том, что на каждой итерации мы удаляем элемент из имеющейся у нас структуры данных, а затем вставляем его в нужную позицию.

На первой итерации мы сортируем два элемента данных. Затем мы расширяем выборку:берм третий элемент и находим для него позицию согласно его значению. Алгоритм выполняется до тех пор, пока все элементы не будут перемещены в правльное положение.

````python
# алгоритм сортировки вставками
def InsertionSort(list):
    for i in range(1, len(list)):
        j = i-1 # индекс предыдущего элемента
        element_next = list[i]  #  сохраняем текущий элемент для вставки
        while (list[j] > element_next) and (j >= 0): #Проверяем, больше ли предыдущий элемент текущего 
            # и Убеждаемся, что не вышли за границы списка
            list[j + 1] = list[j] #  сдвигаем больший элемент вправо
            j = j - 1 # перемещаемся к предыдущему элементу
        list[j+1] = element_next # помещаем текущий элемент на найденное место
    return list        
````

````python
list2 = [25,21,22,24,230,12,26]
print('ДО сортировки вставками:')
print(list2)
InsertionSort(list2)
print('После сортировки вставками:')
print(list2)
````

В основном цикле мы проходим по всему списку. В каждой итерации двумя соседними элементами являются list[j] (текущий элемент) и list[i] (следующий элемент)

В выражениях `(list[j] > element_next) and (j >= 0)` мы сравниваем текущий элемент со следующим

Характеристики алгоритма:
 - Временная сложность: O(n²) в худшем и среднем случаях
 - Пространственная сложность: O(1) — сортировка на месте
 - Устойчивость: алгоритм устойчив (сохраняет порядок равных элементов)
 - Эффективность: хорошо работает на небольших и частично отсортированных массивах

Преимущества:
 - Простота реализации
 - Эффективность на небольших наборах данных
 - Не требует дополнительной памяти
 - Хорошо работает с почти отсортированными данными

### Сортировка слиянием
Производительность не зависит от упорядоченности входных данных. Подобно Map Reduce и другим алгоритмам обработки больших данных, в его основе лежит стратегия "разделяй и властвуй". На этапе "разделения" алгоритм рекурсивно разбивает данные на две части до тех пор, пока размер данных не станет меньше определенного порогового занчения. На этапе слияния алгоритм объединяет данные, пока мы не получим окончательный результат. Логика на рис. 3.7

Принцип работы
Алгоритм состоит из двух основных этапов:

1. Разделение
 - Массив рекурсивно делится пополам
 - Процесс продолжается до тех пор, пока каждый подмассив не будет содержать один элемент
2. Слияние
 - Отсортированные подмассивы объединяются в правильном порядке
 - Происходит сравнение элементов и их последовательное размещение

Временная сложность

Преимущества

 - Требует дополнительной памяти
 - Менее эффективен на малых массивах из-за накладных расходов

````python
def MergeSort(list):
    if len(list)>1:
        mid = len(list)//2 # split list in half
        left = list [:mid]
        right = list[mid:]
        MergeSort(left) #repeats untill lenght of each list is 1
        MergeSort(right)

        a = 0
        b = 0
        c = 0
        while a < len(left) and b <len(right):
            if left[a] < right[b]:
                list[c] = left[a]
                a = a + 1
            else:
                list[c] = right[b]
                b = b + 1
            c = c + 1
        while a < len(left):
            list[c] = left[a]
            a = a + 1
            c = c + 1
        while b < len(right):
            list[c] = right[b]
            b = b + 1
            c = c + 1
    return list     
````

## Сортировка Шелла
В случае с полностью неотсортированным списком размера N алгоритм должен совершить N – 1 полных проходов.

Алгоритм сортировки Шелла доступным языком можно описать так:

Массив делится на подмассивы, каждый из которых состоит из элементов с определённым расстоянием между ними. Обычно это расстояние равно длине массива, делённой на 2.
Массив сортируется с помощью сортировки вставками.
После завершения сортировки промежуток уменьшается вдвое и процесс повторяется.
Этот процесс повторяется до тех пор, пока расстояние не станет равным 1, и массив будет отсортирован.
Сложность
Временная сложность алгоритма сортировки Шелла зависит от выбора шагов. В худшем случае может достигать O(n2), но обычно работает значительно быстрее. лучшая временна́я сложность достигает O(n log2 n)
Пространственная сложность — O(1), так как сортировка выполняется на месте

````python
# из поисковика
def shell_sort(data: list[int]) -> list[int]:
    last_index = len(data)
    step = len(data)//2
    while step > 0:
        for i in range(step, last_index, 1):
            j = i
            delta = j - step
            while delta >= 0 and data[delta] > data[j]:
                data[delta], data[j] = data[j], data[delta]
                j = delta
                delta = j - step
        step //= 2
    return data
````

````python
# алгоритма сортировки Шелла выглядит 
def shellSort(list):
    distance = len(list) // 2
    while distance > 0:
        for i in range(distance, len(list)):
            temp = list[i]
            j = i
    # Сортировка подсписка для текущего значения дистанции
            while j >= distance and list[j - distance] > temp:
                list[j] = list[j - distance]
                j = j-distance
            list[j] = temp
    # Уменьшаем расстояние до следующего элемента
        distance = distance//2
    return list
````

````python
list = [26,17,20,11,23,21,13,18,24,14,12,22,16,15,19,25]
shellSort(list)
print("Sorted array is:\n", list)
````
