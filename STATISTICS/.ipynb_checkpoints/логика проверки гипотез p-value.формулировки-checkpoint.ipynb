Логика проверки гипотез




У вас есть значение, предложенное в нулевой гипотезе. Что ещё нужно, чтобы построить график распределения вероятностей для выборочного среднего?

Чтобы понять, насколько широкое распределение строить, нужно оценить его стандартное отклонение по выборке. Одного размера выборки не хватит. В формуле 
  числитель рассчитывается по всем значениям выборки.
  
  разброс — ESE — оценивается по всей выборке.

при разных уровнях статистической значимости один и тот же эксперимент может как отвергнуть нулевую гипотезу, так и нет.





Итак, уровень значимости (α) — это «граница», которая определяет строгость проверки гипотезы. Чем он ниже, тем сложнее отклонить H0 , но тем меньше вероятность случайной ошибки.

### P-value

 Один из вопросов, интересующих исследователей, — будет ли среднее значение параметра в популяции отличаться от некоторой заданной константы.  
Как вы знаете, для ответа на такие вопросы используют методы статистической проверки гипотез. При этом формулируются два предположения — нулевая гипотеза, в которой параметр считается равным или не отличающимся от заданного значения, и альтернативная гипотеза, которая предполагает отклонение от этого значения. Затем с опорой на данные выборки принимают решение, являются ли наблюдаемые результаты настолько «необычными» для нулевой гипотезы, что её стоит отвергнуть.  
Ключевую роль в этом процессе играет p-value.

Правильно говорить так:   
**«Если нулевая гипотеза верна, то p-value — это вероятность наблюдать значение, равное указанному к нулевой гипотезе или более экстремальное».**

важно понимать:  
**-value указывает на статистическую значимость, а не на практическую ценность результата.** Например, даже незначительные эффекты могут быть статистически значимы в больших выборках. При увеличении выборки тест становится настолько чувствительным, что замечает даже небольшую разницу, которая может не иметь практической ценности. 

Ошибка думать что: малый p-value показывает верность нулевой гипотезы  
Главное заблуждение в том, что отсутствие значимого p-value (p-value > α) доказывает истинность нулевой гипотезы. **Иногда p-value оказывается больше уровня значимости, и это говорит только о том, что у нас недостаточно доказательств, чтобы отклонить нулевую гипотезу (H0)**. Однако это не доказывает, что нулевая гипотеза верна.

Когда сформулированы две гипотезы, происходит следующее:
 - Нулевая гипотеза отвергается в пользу альтернативной, если наблюдаемое на выборке значение оказывается слишком далеко от предполагаемого в нулевой гипотезе.
- Уровень значимости задаёт порог: начиная с какой вероятности наблюдаемое значение считается слишком далёким от предполагаемого, то есть слишком маловероятным.
  
  На практике гипотезы проверяются с помощью расчёта p-value (от англ. p, probability  — «вероятность», value  — «значение») — вероятности получить наблюдаемое значение или большее при условии, что нулевая гипотеза верна. В русскоязычной литературе ещё часто используется термин «p-значение».
